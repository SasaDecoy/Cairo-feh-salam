The helper <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">cliffs_delta()</code> lives here too - it's a Mann-Whitney U effect-size measure used by all three hypothesis sections.</p>

```python
# ── Color palette ──
COLORS = ["#58A6FF","#2A9D8F","#E9C46A","#E86F51","#F4A261","#7B2D8E","#FF6B6B","#4ECDC4"]
BG = "#0D1117"
PANEL = "#161B22"
TXT = "#C9D1D9"
ACCENT = "#58A6FF"
WARN = "#E86F51"
OK = "#2A9D8F"

# ── CRS: all analysis in UTM 36N (meters); web maps only use WGS84 via to_map_crs() ──
CRS_METERS = "EPSG:32636"
CRS_MAP = "EPSG:4326"

def to_map_crs(gdf):
    """Project to WGS84 for Folium / web basemaps. Keep analysis in CRS_METERS everywhere else."""
    if gdf is None or not hasattr(gdf, "to_crs"):
        return gdf
    if gdf.crs is None:
        return gdf
    if str(gdf.crs) == CRS_MAP:
        return gdf
    return gdf.to_crs(CRS_MAP)

px.defaults.template = "plotly_dark"
px.defaults.color_discrete_sequence = COLORS

os.makedirs("Exports", exist_ok=True)

def style(fig, title, h=500):
    """Apply consistent dark styling to any Plotly figure."""
    fig.update_layout(
        title=dict(text=title, font=dict(size=16, color="#E6EDF3")),
        paper_bgcolor=BG, plot_bgcolor=PANEL,
        font=dict(family="Segoe UI, sans-serif", color=TXT, size=12),
        margin=dict(l=50, r=30, t=65, b=50),
        height=h,
        legend=dict(bgcolor="rgba(0,0,0,0)", font_size=11),
    )
    fig.update_xaxes(gridcolor="#21262D", zerolinecolor="#30363D")
    fig.update_yaxes(gridcolor="#21262D", zerolinecolor="#30363D")
    
    # Auto-export the figure
    safe_title = re.sub(r'[^a-zA-Z0-9_\-]', '_', title)
    safe_title = re.sub(r'_+', '_', safe_title).strip('_')
    try:
        fig.write_html(f"Exports/{safe_title}.html", include_plotlyjs="cdn")
    except Exception as e:
        print(f"Warning: Could not export {safe_title} -> {e}")
        
    return fig

def trend(fig, x, y, color=WARN, name="Trend"):
    """Add OLS trendline."""
    m = np.isfinite(x) & np.isfinite(y)
    if m.sum() < 3: return fig
    a, b = np.polyfit(x[m], y[m], 1)
    xx = np.linspace(np.nanmin(x[m]), np.nanmax(x[m]), 100)
    fig.add_trace(go.Scatter(x=xx, y=a*xx+b, mode="lines",
        line=dict(color=color, width=2, dash="dot"), name=name, showlegend=True))
    return fig

def corr(df, x, y):
    d = df[[x,y]]
    return round(d[x].corr(d[y]), 3) if d[x].notna().sum() >= 3 else np.nan

def load_geo(path, crs=None):
    if crs is None:
        crs = CRS_METERS
    df = pd.read_csv(path)
    if "geometry" in df.columns:
        def _p(v):
            if pd.isna(v): return None
            t = str(v).strip()
            if not t or t.lower() in {"nan","none"}: return None
            try: return shapelywkt.loads(t)
            except: return None
        return gpd.GeoDataFrame(df, geometry=df["geometry"].map(_p), crs=crs)
    return gpd.GeoDataFrame(df)

def quintile_labels(s, labels=None):
    """Bin a series into 5 groups."""
    if labels is None:
        labels = ["Very Low","Low","Medium","High","Very High"]
    return pd.qcut(s.rank(method="first"), 5, labels=labels)


# ── Phase 2 addition: Cliff's delta effect size ──
def cliffs_delta(x, y):
    """Effect size for Mann-Whitney U. Returns delta in [-1, 1]."""
    x = np.asarray(list(x)); y = np.asarray(list(y))
    if len(x) == 0 or len(y) == 0: return 0.0
    gt = sum(1 for xi in x for yi in y if xi > yi)
    lt = sum(1 for xi in x for yi in y if xi < yi)
    return (gt - lt) / (len(x) * len(y))

print('palette + cliffs_delta ready')
```

```python
# Metro stations as points in EPSG:32636
metro_pts = gpd.GeoDataFrame(metro.dropna(subset=['lat','lon']),
    geometry=gpd.points_from_xy(metro.dropna(subset=['lat','lon']).lon,
                                  metro.dropna(subset=['lat','lon']).lat),
    crs='EPSG:4326').to_crs('EPSG:32636')

# Aggregate: stations per line
line_counts = metro_pts.groupby('line').size().rename('n_stations').reset_index()
show_df(line_counts, title='Stations per metro line')

# Population density proxy: use Phase 1 cleaned_population hexes
# Pop density is Population_2018 per hex area (hex area constant)
# Attach nearest hex population to each metro station
P = np.column_stack([p1_pop_gdf.geometry.centroid.x, p1_pop_gdf.geometry.centroid.y])
S = np.column_stack([metro_pts.geometry.x, metro_pts.geometry.y])
nn = NearestNeighbors(n_neighbors=5).fit(P)
dists, idxs = nn.kneighbors(S)
metro_pts['hex_pop_2018'] = [p1_pop_gdf.iloc[idx].Population_2018.mean() for idx in idxs]
show_df(metro_pts[['station_en','line','opening_year','hex_pop_2018']].head(5), title='Sample metro stations with population context')
```

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Compute per-line Spearman rank correlation between opening year and average nearby population. Positive correlation suggests newer stations reached denser districts; negative suggests otherwise.</p>

```python
corr_results = []
for line in ['Line 1','Line 2','Line 3']:
    sub = metro_pts[metro_pts.line == line].dropna(subset=['hex_pop_2018','opening_year'])
    if len(sub) < 5: continue
    rho, p = stats.spearmanr(sub['opening_year'], sub['hex_pop_2018'])
    corr_results.append({'line': line, 'n': len(sub),
                         'spearman_rho': round(rho,3), 'p_value': round(p,4),
                         'mean_pop': round(sub['hex_pop_2018'].mean(),0)})
corr_df = pd.DataFrame(corr_results)
show_df(corr_df, title='Q13 · Opening year × nearby hex population (per metro line)',
        highlight_col='spearman_rho')

# Permutation test L3 vs L1
l1_pop = metro_pts[metro_pts.line == 'Line 1'].hex_pop_2018.dropna()
l3_pop = metro_pts[metro_pts.line == 'Line 3'].hex_pop_2018.dropna()
obs_diff = l3_pop.mean() - l1_pop.mean()
combined = np.concatenate([l1_pop, l3_pop])
perms = []
rng = np.random.default_rng(42)
for _ in range(2000):
    rng.shuffle(combined)
    perms.append(combined[:len(l3_pop)].mean() - combined[len(l3_pop):].mean())
p_perm = float(np.mean([abs(p) >= abs(obs_diff) for p in perms]))
show_metrics({'L3 vs L1 mean density diff': f'{obs_diff:+,.0f}',
              'Permutation p-value': f'{p_perm:.4f}'},
             title='Q13 · permutation test')
```

 <h3 style="font-family:'Orbitron',monospace;color:#E9C46A;margin:4px 0 12px;font-size:1.15em;">Do Phase 1's ghost terminals (G1) sit close to the new Phase-3 metro stations, or far from them?</h3>

If they sit far from new metro, the gap persists: Phase 1's structural problem went untouched.</p>
 <p style="margin:6px 0;color:#C9D1D9;line-height:1.55;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#E86F51;">Data path:</strong> merged_G_underused_terminals.csv ⨯ metro_stations.csv (Phase 3, 2014+) via KNN distance in meters</p>

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Load ghost terminals from Phase 1's G1 analysis. Join to Phase 1 terminal coordinates (merged_B) and compute distance to nearest post-2014 metro station.</p>
</div>


```python
ghosts = pd.read_csv(P1_CLEAN / 'merged_G_underused_terminals.csv')
ghosts = ghosts[ghosts['Underused'] == True].copy()
print(f"  Phase 1 ghost terminals: {len(ghosts)}")

# Get geometry from merged_B
tb = pd.read_csv(P1_CLEAN / 'merged_B_terminals_population.csv')
ghost_geom = tb[tb['Terminal_ID'].isin(ghosts['Terminal_ID'])][['Terminal_ID','Terminal_Name','geometry']].drop_duplicates('Terminal_ID')
ghost_gdf = gpd.GeoDataFrame(ghost_geom,
    geometry=ghost_geom['geometry'].apply(wkt.loads), crs='EPSG:32636')
ghost_gdf['x'] = ghost_gdf.geometry.centroid.x
ghost_gdf['y'] = ghost_gdf.geometry.centroid.y

# New metro (post-2014 Line 3 Phase 3B/C/D)
new_metro = metro_pts[(metro_pts.line == 'Line 3') & (metro_pts.opening_year >= 2014)].copy()
print(f"  New (post-2014) metro stations: {len(new_metro)}")

# Distance from each ghost to nearest new metro station
NM = np.column_stack([new_metro.geometry.x, new_metro.geometry.y])
G = np.column_stack([ghost_gdf['x'], ghost_gdf['y']])
nn = NearestNeighbors(n_neighbors=1).fit(NM)
dist_m, idx = nn.kneighbors(G)
ghost_gdf['dist_to_new_metro_m'] = dist_m[:,0]
ghost_gdf['nearest_new_metro']   = new_metro.iloc[idx[:,0]]['station_en'].values
```

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Classify ghost terminals by distance bucket and count how many are 'reached' vs 'stranded'.</p>

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Build a timeline of cumulative metro kilometers by taking each line's station opening years and summing inter-station distances in the order they opened. Phase 1 terminal count is held constant at 280 (no Phase 1 equivalent for 'new' bus terminals).</p>

```python
metro_timeline = metro_pts[metro_pts['opening_year'].notna()].sort_values('opening_year').copy()
# Approximate inter-station distance as ~1.2 km (typical Cairo metro stop spacing)
AVG_STOP_KM = 1.2
metro_timeline['cumulative_km'] = (metro_timeline.reset_index().index + 1) * AVG_STOP_KM

# Yearly cumulative km
years = sorted(metro_timeline['opening_year'].dropna().unique().astype(int))
ratio_rows = []
for y in years:
    km = metro_timeline[metro_timeline.opening_year <= y].shape[0] * AVG_STOP_KM
    bus_terms = 280
    ratio_rows.append({'year': y, 'metro_km': km, 'bus_terminals': bus_terms, 'ratio_km_per_terminal': km / bus_terms})
ratio_ts = pd.DataFrame(ratio_rows)
print(ratio_ts.to_string(index=False))
```

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Use the cagr_2006_2017 column already computed in districts_wide. Top 15 by growth rate. Attach nearest-station count using a 2 km buffer around each district's population centroid (proxy for district centroid since we lack polygons).</p>

```python
# Q16 · Fastest-growing districts AND their new-mode coverage.
# Real question: fast growth without coverage = Masari's forward pipeline.
dw_q16 = districts_w.copy()
dw_q16['cagr_pct'] = dw_q16['cagr_2006_2017'] * 100
dw_q16 = dw_q16.dropna(subset=['cagr_pct'])

# For each district, count new-mode stations (Metro L3 post-2014 + LRT + BRT) within 3 km
# of its REAL Nominatim centroid (governorate-fallback for the 26 that Nominatim missed).
new_modes_gdf = integrated[
    integrated.geometry.notna() &
    integrated['source_label'].isin(['S3_metro','S4_lrt','S5_brt'])
].copy()

def _count_new_modes(row, radius_m=3000):
    pt = gpd.GeoDataFrame([{'g': Point(row['centroid_lon'], row['centroid_lat'])}],
        geometry='g', crs='EPSG:4326').to_crs('EPSG:32636').geometry.iloc[0]
    dx = new_modes_gdf.geometry.x - pt.x
    dy = new_modes_gdf.geometry.y - pt.y
    return int(((dx**2 + dy**2) ** 0.5 < radius_m).sum())

# Only count once per unique centroid (saves time when gov-fallback districts share a point)
uc = dw_q16[['centroid_lat','centroid_lon']].drop_duplicates().reset_index(drop=True)
uc['n_new_modes'] = uc.apply(_count_new_modes, axis=1)
dw_q16 = dw_q16.merge(uc, on=['centroid_lat','centroid_lon'], how='left')

top15 = dw_q16.nlargest(15, 'cagr_pct')[[
    'name','governorate','pop_2017','pop_2023','cagr_pct','n_new_modes','centroid_source']].copy()

show_df(top15, title='Q16 · top 15 fastest-growing districts and their new-mode coverage',
        caption='n_new_modes = stations of any new formal mode (Metro L3 post-2014 / LRT / BRT) within 3 km.',
        highlight_col='n_new_modes')

uncovered = (top15['n_new_modes'] == 0).sum()
show_metrics({
    'Top 15 fastest-growing districts':            15,
    'With ANY new-mode station within 3 km':       int(15 - uncovered),
    'With ZERO new-mode station within 3 km':      int(uncovered),
    'Mean CAGR of uncovered districts':            (
        f"{top15.loc[top15.n_new_modes == 0, 'cagr_pct'].mean():.1f}%"
        if uncovered else 'n/a'
    ),
}, title='Q16 · coverage shortfall at the growth frontier',
    highlight={'With ZERO new-mode station within 3 km'})
```

```python
q4 = pd.read_csv(P1_CLEAN / 'q4_formal_vs_informal.csv')
q4['Informal_Share_Daily'] = 1 - q4['Formal_Share_Daily'].fillna(0)
q4 = q4.dropna(subset=['geometry','Informal_Share_Daily']).copy()
q4_gdf = gpd.GeoDataFrame(q4, geometry=q4['geometry'].apply(wkt.loads), crs='EPSG:32636')

# Nearest hex population for each stop
SP = np.column_stack([p1_pop_gdf.geometry.centroid.x, p1_pop_gdf.geometry.centroid.y])
QP = np.column_stack([q4_gdf.geometry.x, q4_gdf.geometry.y])
nn = NearestNeighbors(n_neighbors=1).fit(SP)
_, idx = nn.kneighbors(QP)
q4_gdf['hex_pop'] = p1_pop_gdf.iloc[idx[:,0]]['Population_2018'].values
q4_gdf = q4_gdf.dropna(subset=['hex_pop','Informal_Share_Daily'])
q4_gdf = q4_gdf[q4_gdf['hex_pop'] > 0]
print(f"  Stops with pop + share: {len(q4_gdf)}")
print(f"  Mean informal share: {q4_gdf.Informal_Share_Daily.mean():.3f}")

# Regression
X = np.log10(q4_gdf['hex_pop'].values).reshape(-1,1)
y = q4_gdf['Informal_Share_Daily'].values
reg = LinearRegression().fit(X, y)
rho, p = stats.spearmanr(q4_gdf['hex_pop'], q4_gdf['Informal_Share_Daily'])
print(f"  Spearman rho on (log_pop, informal_share): {rho:.3f}  p={p:.2e}")
print(f"  OLS slope (log10 pop → informal share): {reg.coef_[0]:.4f}")

# ── S7 integration: vehicles-per-capita per governorate ──
# Bring in the motor-vehicle total (WB-derived national × S6 governorate population share)
# and join it to each Phase-1 stop via its hex's governorate. Lets us ask a sharper
# question: does the informal share drop where private motorization is high?
veh = vehicles[vehicles.vehicle_type == 'motor_vehicles_total'][['governorate','year','count']].copy()
veh_2023 = veh[veh.year == 2023].rename(columns={'count':'veh_total'})[['governorate','veh_total']]

dw_pop = districts_w.copy()
dw_pop['gov_short'] = dw_pop['governorate'].str.extract(r'\[ ([^\]]+) \]')[0].fillna(dw_pop['governorate'])
gov_pop = dw_pop.groupby('gov_short')['pop_2023'].sum().reset_index()
gov_pop = gov_pop.merge(veh_2023, left_on='gov_short', right_on='governorate', how='left')
gov_pop['vehicles_per_capita'] = gov_pop['veh_total'] / gov_pop['pop_2023']
print('\n  Governorate motorization (vehicles per capita, 2023):')
print(gov_pop[['gov_short','pop_2023','veh_total','vehicles_per_capita']].to_string(index=False))
```

```python
# Q18b · Does each new mode close any of Phase 1's structural gaps?
# For each of Phase 1's four documented gaps (G1 ghost terminals, G2 empty-return, G3 vehicle-route mismatch,
# G4 top-quartile underserved hexes) and each of the three new modes (Metro L3 post-2014, LRT, BRT),
# compute the % of gap instances within 2 km of a station of that mode.
from shapely import wkt as _wkt

# Gap geographies
g1_src = pd.read_csv(P1_CLEAN / 'merged_G_underused_terminals.csv')
g1_src = g1_src[g1_src['Underused'] == True]
tb = pd.read_csv(P1_CLEAN / 'merged_B_terminals_population.csv')
g1_geom = tb[tb['Terminal_ID'].isin(g1_src['Terminal_ID'])][['Terminal_ID','geometry']].drop_duplicates('Terminal_ID')
g1_gdf = gpd.GeoDataFrame(g1_geom, geometry=g1_geom['geometry'].apply(_wkt.loads), crs='EPSG:32636')
g1_gdf['x'] = g1_gdf.geometry.centroid.x; g1_gdf['y'] = g1_gdf.geometry.centroid.y

g2_flow = pd.read_csv(P1_CLEAN / 'merged_D_terminal_flow_ratios.csv')
g2_flow = g2_flow[g2_flow['Empty_Return_Index'] >= 0.9]
g2_geom = tb[tb['Terminal_ID'].isin(g2_flow['Terminal_ID'])][['Terminal_ID','geometry']].drop_duplicates('Terminal_ID')
g2_gdf = gpd.GeoDataFrame(g2_geom, geometry=g2_geom['geometry'].apply(_wkt.loads), crs='EPSG:32636')
g2_gdf['x'] = g2_gdf.geometry.centroid.x; g2_gdf['y'] = g2_gdf.geometry.centroid.y

g3_src = pd.read_csv(P1_CLEAN / 'merged_F_routes_per_terminal.csv')
g3_src = g3_src[g3_src['Avg_Route_Len_Origin'] > 50]
g3_geom = tb[tb['Terminal_ID'].isin(g3_src['Terminal_ID'])][['Terminal_ID','geometry']].drop_duplicates('Terminal_ID')
g3_gdf = gpd.GeoDataFrame(g3_geom, geometry=g3_geom['geometry'].apply(_wkt.loads), crs='EPSG:32636')
g3_gdf['x'] = g3_gdf.geometry.centroid.x; g3_gdf['y'] = g3_gdf.geometry.centroid.y

q9 = pd.read_csv(P1_CLEAN / 'q9_underserved.csv')
q9_top = q9[q9['Underserved_Score'] >= q9['Underserved_Score'].quantile(0.75)]
g4_gdf = p1_pop_gdf[p1_pop_gdf['H3_Hex_ID'].isin(q9_top['H3_Hex_ID'])].copy()
g4_gdf['x'] = g4_gdf.geometry.centroid.x; g4_gdf['y'] = g4_gdf.geometry.centroid.y

# Mode station sets
new_metro = metro_pts[(metro_pts.line == 'Line 3') & (metro_pts.opening_year >= 2014)].copy()
lrt_pts = gpd.GeoDataFrame(lrt.dropna(subset=['lat','lon']).copy(),
    geometry=gpd.points_from_xy(lrt.dropna(subset=['lat','lon']).lon, lrt.dropna(subset=['lat','lon']).lat),
    crs='EPSG:4326').to_crs('EPSG:32636')
brt_pts_all = gpd.GeoDataFrame(brt.dropna(subset=['lat','lon']).copy(),
    geometry=gpd.points_from_xy(brt.dropna(subset=['lat','lon']).lon, brt.dropna(subset=['lat','lon']).lat),
    crs='EPSG:4326').to_crs('EPSG:32636')

modes = {'Metro L3 (post-2014)': new_metro, 'LRT (2022)': lrt_pts, 'BRT (2025)': brt_pts_all}
gaps  = {'G1 ghosts': g1_gdf, 'G2 empty-return': g2_gdf, 'G3 vehicle mismatch': g3_gdf, 'G4 underserved hexes': g4_gdf}

# Coverage matrix (% of gap instances within 2 km of a mode station)
RADIUS_M = 2000
matrix = pd.DataFrame(index=list(modes.keys()), columns=list(gaps.keys()), dtype=float)
for mode_name, m_gdf in modes.items():
    if len(m_gdf) == 0:
        matrix.loc[mode_name] = 0; continue
    M = np.column_stack([m_gdf.geometry.x, m_gdf.geometry.y])
    nn = NearestNeighbors(n_neighbors=1).fit(M)
    for gap_name, g_gdf in gaps.items():
        G = np.column_stack([g_gdf['x'], g_gdf['y']])
        dist, _ = nn.kneighbors(G)
        matrix.loc[mode_name, gap_name] = (dist[:,0] < RADIUS_M).sum() / len(g_gdf) * 100

show_df(matrix.round(1).reset_index().rename(columns={'index':'mode'}),
        title='Q18b · Do new modes close Phase-1 gaps? (% of gap within 2 km)',
        caption='Row-major: which mode closed most gaps. Column-major: which gaps went unfilled.')

# Heatmap
fig = go.Figure(go.Heatmap(
    z=matrix.values.astype(float), x=list(matrix.columns), y=list(matrix.index),
    colorscale=[[0,'#0D1117'],[0.25,'#E86F51'],[0.6,'#E9C46A'],[1,'#2A9D8F']],
    colorbar=dict(title='% covered', len=0.7),
    text=[[f'{v:.0f}%' for v in row] for row in matrix.values],
    texttemplate='%{text}', textfont=dict(color='#0D1117', size=16),
    hovertemplate='%{y} covers %{x}<br>%{z:.1f}% of instances<extra></extra>',
    zmin=0, zmax=100,
))
fig.update_layout(template='plotly_dark', paper_bgcolor=BG, plot_bgcolor=BG,
    title=dict(text='Q18b · Do the new modes close Phase-1 gaps?', font=dict(color='#E9C46A')),
    xaxis=dict(color=TXT, tickangle=-10), yaxis=dict(color=TXT, autorange='reversed'),
    font=dict(color=TXT), height=400, width=1100, margin=dict(l=220))
fig.show()
```

```python
# TfC bus/metro GTFS — route counts per agency (single bundle).

def load_routes_agency(zip_path):
    out = {}
    with zipfile.ZipFile(zip_path) as z:
        for fn in ['routes.txt','agency.txt']:
            try:
                with z.open(fn) as f:
                    out[fn.replace('.txt','')] = pd.read_csv(f)
            except KeyError:
                out[fn.replace('.txt','')] = pd.DataFrame()
    return out

RAW = Path('RawData')
bm = load_routes_agency(RAW / 'gtfs_bus_metro.zip')
cov = bm['routes'].merge(bm['agency'], on='agency_id', how='left').groupby('agency_name').size().rename('bus_metro_routes').reset_index()
cov['bus_metro_routes'] = cov['bus_metro_routes'].astype(int)
cov = cov.sort_values('bus_metro_routes', ascending=False)
print(cov.to_string(index=False))
print(f"\n  Agencies in bundle: {len(cov)}")
```

 <h3 style="font-family:'Orbitron',monospace;color:#E9C46A;margin:4px 0 12px;font-size:1.15em;">Which mode - formal or informal - offers the better fare per kilometer, and is the difference enough to explain Layla's choice?</h3>
 <p style="margin:6px 0;color:#C9D1D9;line-height:1.55;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#2A9D8F;">Why it matters:</strong> Layla's 83-minute commute is partly an economic decision.  <p style="margin:6px 0;color:#C9D1D9;line-height:1.55;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#E86F51;">Data path:</strong> q12_vehicle_type_route_length.csv (Vehicle_Type × Route_Length_km) + GTFS fare_attributes.csv</p>

<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Compute fare/km distribution by vehicle type. Map Phase 1 Vehicle_Type categories to formal/informal buckets. Overlay a strip plot.</p>

```python
# Use GTFS fare_attributes for REAL formal fares + Phase 1 q12 for informal vehicle-type fares
fares = pd.read_csv(P2_CLEAN / 'gtfs_fare_attributes.csv')
print('GTFS fare_attributes:')
print(fares.to_string(index=False))

# Compute route-level fares: join fare_rules → fare_attributes, attach route_id
# Not all GTFS bundles have route-level fares; if absent, fall back to agency-average
gtfs_fr = pd.read_csv(Path('RawData') / 'gtfs_bus_metro/fare_rules.txt', low_memory=False) if (Path('RawData') / 'gtfs_bus_metro/fare_rules.txt').exists() else pd.DataFrame()
print(f"fare_rules rows: {len(gtfs_fr)}")

gtfs_routes_full = pd.read_csv(P2_CLEAN / 'gtfs_routes.csv')
# Pseudo route-length from shapes — sum consecutive haversine along shape_pt_sequence
sh = pd.read_csv(P2_CLEAN / 'gtfs_shapes.csv')
sh_gdf = gpd.GeoDataFrame(sh, geometry=gpd.points_from_xy(sh.shape_pt_lon, sh.shape_pt_lat),
                            crs='EPSG:4326').to_crs('EPSG:32636')
sh_gdf = sh_gdf.sort_values(['shape_id','shape_pt_sequence'])
def shape_len(group):
    xy = np.column_stack([group.geometry.x, group.geometry.y])
    d = np.hypot(np.diff(xy[:,0]), np.diff(xy[:,1])).sum() / 1000.0
    return d
shape_lengths = sh_gdf.groupby('shape_id').apply(shape_len, include_groups=False).rename('length_km').reset_index()
trips = pd.read_csv(P2_CLEAN / 'gtfs_trips.csv')
route_shapes = trips.dropna(subset=['shape_id']).groupby('route_id')['shape_id'].first().reset_index()
route_lengths = route_shapes.merge(shape_lengths, on='shape_id', how='left')
routes_out = gtfs_routes_full.merge(route_lengths, on='route_id', how='left')

# Merge fare: join on fare_rules route_id (if exists) else fall back to mean fare
if 'route_id' in gtfs_fr.columns and len(gtfs_fr):
    route_fares = gtfs_fr.merge(fares, on='fare_id', how='left').groupby('route_id')['price'].mean().reset_index().rename(columns={'price':'fare_egp'})
    routes_out = routes_out.merge(route_fares, on='route_id', how='left')
else:
    routes_out['fare_egp'] = fares['price'].mean()

routes_out['fare_per_km'] = routes_out['fare_egp'] / routes_out['length_km']
routes_out = routes_out[routes_out['length_km'] > 0].dropna(subset=['fare_per_km'])

# Join Phase 1 Q12 (informal-only vehicle types)
q12 = pd.read_csv(P1_CLEAN / 'q12_vehicle_type_route_length.csv')
# Approximate informal fares from route length — assume 0.5 EGP/km baseline for informal (regulator cap)
informal = q12.copy()
informal['fare_per_km'] = 7.0 / informal['Route_Length_km'].clip(lower=1)
informal['vehicle_type'] = informal['Vehicle_Type']
informal['source_bucket'] = 'informal'

# Tag formal routes by route_type
route_type_map = {0:'Tram',1:'Subway (Metro)',3:'Bus',7:'Funicular',11:'Trolleybus'}
routes_out['vehicle_type'] = routes_out.get('route_type', pd.Series(dtype=int)).map(route_type_map).fillna('Bus')
routes_out['source_bucket'] = 'formal'

combined = pd.concat([
    routes_out[['vehicle_type','fare_per_km','source_bucket']],
    informal[['vehicle_type','fare_per_km','source_bucket']],
], ignore_index=True)

print('\n  Fare per km summary:')
print(combined.groupby(['source_bucket','vehicle_type'])['fare_per_km'].describe().round(2).to_string())
```

 <h3 style="font-family:'Orbitron',monospace;color:#E9C46A;margin:4px 0 12px;font-size:1.15em;">Given a district's population, how many metro stations should it have - and how far is reality from that?</h3>
 <p style="margin:6px 0;color:#C9D1D9;line-height:1.55;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#2A9D8F;">Why it matters:</strong> The residual is the coverage-need gap per district. Positive residual = over-served, negative = under-served. This is the single best diagnostic for Masari's market map.</p>
 <p style="margin:6px 0;color:#C9D1D9;line-height:1.55;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#E86F51;">Data path:</strong> metro_stations.csv counts per district ⨯ districts_wide population → OLS residual</p>

<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Share Tech Mono',monospace;color:#58A6FF;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">⟐ CODE EXPLANATION</span>
 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Approximate each metro station's district by matching against the district's governorate. For each district, count metro stations within a 3 km buffer around its nominal centroid (proxy). Fit OLS population → stations; residual is the gap.</p>
</div>


```python
# Q22 · coverage residual per district, using REAL Nominatim centroids from S6.
from sklearn.linear_model import LinearRegression

dw2 = districts_w.copy()
dw2 = dw2[dw2.pop_2023.notna() & (dw2.pop_2023 > 0)].copy()
# Use real centroids from S6's Nominatim geocoding (fallback to gov centroid is already filled in)
dw2['cx'] = dw2['centroid_lon']
dw2['cy'] = dw2['centroid_lat']

# Count metro stations within 3 km of each district's real centroid (not gov proxy).
metro_pts_wgs = metro_pts.to_crs('EPSG:4326')

def count_within(row, radius_m=3000):
    p = Point(row['cx'], row['cy'])
    gdf = gpd.GeoDataFrame([{'g': p}], geometry='g', crs='EPSG:4326').to_crs('EPSG:32636')
    px, py = gdf.geometry.iloc[0].x, gdf.geometry.iloc[0].y
    dx = metro_pts.geometry.x - px
    dy = metro_pts.geometry.y - py
    return int(((dx**2 + dy**2) ** 0.5 < radius_m).sum())

# Reuse if same centroid (Nominatim misses collapse to gov centroid → cache by (cx,cy))
unique_centroids = dw2[['cx','cy']].drop_duplicates().reset_index(drop=True)
unique_centroids['n_stations'] = unique_centroids.apply(count_within, axis=1)
dw2 = dw2.merge(unique_centroids, on=['cx','cy'], how='left')

X = dw2[['pop_2023']].values; y = dw2['n_stations'].values
reg = LinearRegression().fit(X, y)
pred = reg.predict(X)
dw2['residual'] = y - pred

show_metrics({
    'Districts with pop_2023':     len(dw2),
    'Unique centroid locations':   dw2[['cx','cy']].drop_duplicates().shape[0],
    'OLS slope (per person)':      f'{reg.coef_[0]:.2e}',
    'OLS intercept':                f'{reg.intercept_:.2f}',
    'Residual p25 / p75':          f'{np.percentile(dw2.residual, 25):+.2f} / {np.percentile(dw2.residual, 75):+.2f}',
}, title='Q22 · OLS regression on real district centroids', highlight={'Unique centroid locations'})

show_df(
    dw2.sort_values('residual').head(10)[['name','governorate','pop_2023','n_stations','residual','centroid_source']],
    title='Q22 · top 10 under-served districts (most-negative residual)',
    caption='n_stations = metro stations within 3 km of the district\'s real (Nominatim or gov-fallback) centroid.',
    highlight_col='residual'
)
```

 <span style="font-family:'Orbitron',monospace;color:#58A6FF;font-size:.7em;letter-spacing:.3em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Q24</span>
 <h3 style="font-family:'Orbitron',monospace;color:#E9C46A;margin:4px 0 12px;font-size:1.15em;">If we cluster Greater Cairo districts by coverage, density, and growth, which cluster is Masari's primary market?</h3>
 <p style="margin:6px 0;color:#C9D1D9;line-height:1.55;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#2A9D8F;">Why it matters:</strong> Masari cannot serve all of Cairo simultaneously. Rigorous segmentation (5 features, K-Means) identifies the smallest cluster that captures the largest addressable informal demand.</p>
 <p style="margin:6px 0;color:#C9D1D9;line-height:1.55;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#E86F51;">Data path:</strong> districts_wide.csv (5 standardized features: density, CAGR, metro residual, informal share, gov share)</p>

<div style="background:rgba(123,45,142,0.08);border:1.5px dashed #7B2D8E;padding:22px 24px;border-radius:8px;margin:20px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Share Tech Mono',monospace;color:#7B2D8E;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">⟁ METHODOLOGY ENHANCEMENT</span>
 <p style="margin:10px 0 0;color:#C9D1D9;line-height:1.65;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong>AI Technique 2 · K-Means Clustering.</strong> Five features per district, StandardScaler'd, elbow + silhouette across k∈[2,8], Adjusted Rand Index across 5 random seeds for stability. If ARI < 0.6, report it honestly and bump <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">n_init=50</code>. The chosen k produces named clusters (the highest informal-to-station gap cluster, Wealthy Formal, Peripheral Growth, etc.) for downstream market sizing.</p>
</div>


<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Share Tech Mono',monospace;color:#58A6FF;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">⟐ CODE EXPLANATION</span>
 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Build the feature matrix. For each district: pop density proxy (pop_2023), CAGR, nearest-station count from Q22, informal demand intensity (from Q4 stops), governorate share encoded.</p>
</div>


```python
# Q24 feature matrix built on REAL district centroids.
# Earlier version used 3 gov-centroid proxies → all districts in Cairo had the same n_stations / n_informal.
# With Nominatim centroids (42 real + 26 gov-fallback = 37 unique), per-district features now vary.
feat = dw2.copy()
feat['log_pop_2023'] = np.log10(feat['pop_2023'] + 1)
feat['cagr_pct']     = feat['cagr_2006_2017'].fillna(0) * 100

# Per-district informal stop count: use each district's real centroid with a 3 km buffer.
if 'q4_gdf' in dir():
    q4_informal_ok = q4_gdf[q4_gdf.Informal_Share_Daily.notna()].copy()
    uc = feat[['cx','cy']].drop_duplicates().reset_index(drop=True)
    uc_proj = gpd.GeoDataFrame(uc,
        geometry=gpd.points_from_xy(uc.cx, uc.cy), crs='EPSG:4326').to_crs('EPSG:32636')
    Q = np.column_stack([q4_informal_ok.geometry.x, q4_informal_ok.geometry.y])
    counts = []
    for _, r in uc_proj.iterrows():
        x0, y0 = r.geometry.x, r.geometry.y
        mask = ((Q[:,0]-x0)**2 + (Q[:,1]-y0)**2) ** 0.5 < 3000
        counts.append(int(mask.sum()))
    uc_proj['n_informal_stops'] = counts
    feat = feat.merge(uc_proj[['cx','cy','n_informal_stops']], on=['cx','cy'], how='left')
else:
    feat['n_informal_stops'] = 0

feat['is_cairo'] = feat.governorate.str.contains('Cairo', na=False).astype(int)
feat['is_giza']  = feat.governorate.str.contains('Giza',  na=False).astype(int)

X_cols = ['log_pop_2023','cagr_pct','n_stations','n_informal_stops','is_cairo']
X = feat[X_cols].fillna(0).to_numpy()

show_metrics({
    'Feature matrix shape':         f'{X.shape[0]} × {X.shape[1]}',
    'Unique centroid locations':    feat[['cx','cy']].drop_duplicates().shape[0],
    'Unique n_stations values':     feat.n_stations.nunique(),
    'Unique n_informal values':     feat.n_informal_stops.nunique(),
}, title='Q24 · feature matrix (post-Nominatim upgrade)',
    highlight={'Unique n_stations values','Unique n_informal values'})

show_df(feat[X_cols].describe().round(2).reset_index().rename(columns={'index':'stat'}),
        title='Q24 · per-feature summary statistics', caption='Before Nominatim: n_stations had 3 unique values. Now it has many.')
```

<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Share Tech Mono',monospace;color:#58A6FF;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">⟐ CODE EXPLANATION</span>
 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Elbow + silhouette over k∈[2,8]. Pick optimal k by highest silhouette (tie-break by lowest inertia).</p>
</div>


```python
scaler = StandardScaler()
Xs = scaler.fit_transform(X)

ks = list(range(2, 9))
inertias, sils = [], []
for k in ks:
    km = KMeans(n_clusters=k, n_init=20, random_state=0).fit(Xs)
    inertias.append(km.inertia_)
    sils.append(silhouette_score(Xs, km.labels_))
elbow_df = pd.DataFrame({'k': ks, 'inertia': inertias, 'silhouette': sils})
print(elbow_df.to_string(index=False))

# Pick k=4 explicitly: we need at least 4 clusters to distinguish the four segments
# of interest (Formal core, Dense Informal, Peripheral Growth, Wealthy Fringe).
# Silhouette-argmax picks k=2 which under-segments. We log that trade-off.
best_k = 4
print(f"  Silhouette-argmax would pick k={ks[int(np.argmax(sils))]} but we force k={best_k} for policy-relevant granularity.")

fig = make_subplots(rows=1, cols=2, subplot_titles=['Elbow (inertia)','Silhouette'])
fig.add_trace(go.Scatter(x=ks, y=inertias, mode='lines+markers',
    line=dict(color='#58A6FF'), marker=dict(size=9)), row=1, col=1)
fig.add_trace(go.Scatter(x=ks, y=sils, mode='lines+markers',
    line=dict(color='#E9C46A'), marker=dict(size=9)), row=1, col=2)
fig.add_vline(x=best_k, line_dash='dot', line_color='#E86F51')
fig.update_layout(template='plotly_dark', paper_bgcolor=BG, plot_bgcolor=BG,
    title=dict(text=f'K-Means model selection (forced k = {best_k}, policy granularity)',
               font=dict(color='#E9C46A')),
    font=dict(color=TXT), height=400, width=1050, showlegend=False)
fig.show()
```

<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Share Tech Mono',monospace;color:#58A6FF;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">⟐ CODE EXPLANATION</span>
 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Stability check: 5 different random seeds, Adjusted Rand Index across them. ARI > 0.6 = stable clustering.</p>
</div>


```python
base_labels = KMeans(n_clusters=best_k, n_init=50, random_state=0).fit_predict(Xs)
aris = []
for seed in [1,2,3,4,5]:
    labels = KMeans(n_clusters=best_k, n_init=50, random_state=seed).fit_predict(Xs)
    aris.append(adjusted_rand_score(base_labels, labels))
mean_ari = float(np.mean(aris))
print(f"  ARI across 5 seeds: {aris}")
print(f"  mean ARI: {mean_ari:.3f}  (>0.6 = stable)")

feat['cluster'] = base_labels
# Profile clusters
profile = feat.groupby('cluster').agg(
    n_districts=('name','count'),
    median_pop=('pop_2023','median'),
    median_cagr=('cagr_pct','median'),
    median_stations=('n_stations','median'),
    median_informal=('n_informal_stops','median'),
    cairo_pct=('is_cairo','mean'),
).round(2)
print(profile.to_string())
```

<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Share Tech Mono',monospace;color:#58A6FF;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">⟐ CODE EXPLANATION</span>
 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Name each cluster using its profile, and visualize as 2-panel stacked bar (cluster × feature z-scores + cluster sizes).</p>
</div>


```python
# Name each cluster by its profile vs the overall median
med_pop   = profile.median_pop.median()
med_cagr  = profile.median_cagr.median()
med_stn   = profile.median_stations.median()
med_inf   = profile.median_informal.median()

cluster_names = {}
for cid, row in profile.iterrows():
    high_pop  = row.median_pop > med_pop
    high_cagr = row.median_cagr > med_cagr
    high_stn  = row.median_stations > med_stn
    high_inf  = row.median_informal > med_inf
    is_cairo  = row.cairo_pct > 0.5
    # Priority ordering
    if high_inf and not high_stn and high_pop:
        cluster_names[cid] = 'Dense Informal Core'
    elif high_stn and is_cairo:
        cluster_names[cid] = 'Formal-Served Core'
    elif high_cagr and not is_cairo:
        cluster_names[cid] = 'Peripheral Growth'
    elif not high_pop and high_stn:
        cluster_names[cid] = 'Low-Density Formal Fringe'
    elif not high_pop and not high_stn:
        cluster_names[cid] = 'Low-Activity Outskirts'
    else:
        cluster_names[cid] = f'Mixed (cluster {cid})'

# Ensure uniqueness
seen = {}
for cid, nm in list(cluster_names.items()):
    if nm in seen.values():
        cluster_names[cid] = f'{nm} ({cid})'
    seen[cid] = cluster_names[cid]

feat['cluster_name'] = feat['cluster'].map(cluster_names)
print(f"Cluster naming (k={best_k}):")
for cid, nm in cluster_names.items():
    print(f"  {cid}: {nm}  (n={int(profile.loc[cid].n_districts)})")

# 2-panel: left = size per cluster, right = feature z-score heat
sizes = feat.groupby('cluster_name').size().sort_values(ascending=True)
fig = make_subplots(rows=1, cols=2, subplot_titles=['Cluster size (districts)','Mean feature z-score'],
                    column_widths=[0.45,0.55])
palette4 = ['#58A6FF','#2A9D8F','#E9C46A','#E86F51','#7B2D8E','#F4A261']
fig.add_trace(go.Bar(y=sizes.index, x=sizes.values, orientation='h',
    marker_color=palette4[:len(sizes)]), row=1, col=1)
z = pd.DataFrame(Xs, columns=X_cols); z['cluster'] = base_labels
zp = z.groupby('cluster').mean()
zp.index = [cluster_names[i] for i in zp.index]
fig.add_trace(go.Heatmap(z=zp.values, x=zp.columns, y=zp.index,
    colorscale=[[0,'#0D1117'],[0.5,'#58A6FF'],[1,'#E9C46A']], colorbar=dict(len=.5)), row=1, col=2)
fig.update_layout(template='plotly_dark', paper_bgcolor=BG, plot_bgcolor=BG,
    title=dict(text=f'Q24 · K-Means segmentation (k={best_k}, mean ARI={mean_ari:.2f})',
               font=dict(color='#E9C46A')),
    font=dict(color=TXT), height=480, width=1100, showlegend=False)
fig.show()
```

 <h2 style="font-family:'Orbitron',monospace;font-size:1.6em;color:#E9C46A;margin:6px 0;">HYPOTHESIS H1 · COVERAGE-NEED MISMATCH + MORAN'S I</h2>

<div style="background:rgba(123,45,142,.06);border:1.5px solid rgba(123,45,142,.35);border-radius:10px;padding:18px 22px;margin:14px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Orbitron',monospace;color:#7B2D8E;font-size:.7em;letter-spacing:.3em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">HYPOTHESIS H1</span>
 <p style="margin:8px 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#58A6FF;">H₀:</strong> Districts in every density tertile receive the same rate of new-mode stations per capita.</p>
 <p style="margin:8px 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#2A9D8F;">H₁:</strong> Dense districts receive systematically fewer new-mode stations per capita than medium/low-density districts.</p>
 <p style="margin:8px 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#E9C46A;">Test:</strong> Kruskal-Wallis H (non-parametric ANOVA) across three density tertiles, followed by pairwise Mann-Whitney U with Bonferroni correction</p>
 <p style="margin:8px 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#E86F51;">Effect size:</strong> ε² (epsilon-squared) for Kruskal-Wallis · Cliff's δ for pairwise tests</p>
</div>


<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Share Tech Mono',monospace;color:#58A6FF;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">⟐ CODE EXPLANATION</span>
 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Tertile-split districts by 2023 population. Compute new-mode stations (metro+LRT+BRT) per capita per district. Run Kruskal-Wallis + post-hoc Mann-Whitney U.</p>
</div>


```python
d = feat.copy()
d['stations_per_100k'] = (d['n_stations'] / d['pop_2023'] * 100_000).replace([np.inf, -np.inf], np.nan).fillna(0)
d['density_tertile'] = pd.qcut(d['pop_2023'], q=3, labels=['low','med','high'])

groups = [d[d.density_tertile == t]['stations_per_100k'].values for t in ['low','med','high']]
h_stat, h_p = stats.kruskal(*groups)
N = sum(len(g) for g in groups)
eps2 = (h_stat - (len(groups) - 1)) / (N - len(groups))

show_metrics({'Kruskal-Wallis H': round(h_stat, 3),
              'p-value':          f'{h_p:.2e}',
              'ε² (effect size)': round(eps2, 3)},
             title='H1 · overall test', highlight={'ε² (effect size)'})

tertile_stats = d.groupby('density_tertile', observed=True)['stations_per_100k'].describe().round(2)
show_df(tertile_stats.reset_index(), title='H1 · stations per 100k residents by density tertile',
        highlight_col='50%')

from itertools import combinations
pairs = list(combinations(['low','med','high'], 2))
n_comp = len(pairs)
pw = []
for a, b in pairs:
    u, p = stats.mannwhitneyu(d[d.density_tertile == a].stations_per_100k,
                               d[d.density_tertile == b].stations_per_100k, alternative='two-sided')
    delta = cliffs_delta(d[d.density_tertile == a].stations_per_100k,
                          d[d.density_tertile == b].stations_per_100k)
    pw.append({'pair': f'{a} vs {b}', 'U': u, 'p': round(p, 6),
               'p_bonferroni': round(min(1.0, p*n_comp), 6), 'cliffs_delta': round(delta, 3)})
show_df(pd.DataFrame(pw), title='H1 · pairwise Mann-Whitney U (Bonferroni)',
        highlight_col='cliffs_delta')
```

 <p style="margin:10px 0 0;color:#C9D1D9;line-height:1.65;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong>AI Technique 3 · Moran's I · Spatial Autocorrelation.</strong> (Upgraded: now uses real Nominatim district centroids from S6, not the earlier 3-gov-centroid proxy - makes the KNN weights graph meaningful.) Kruskal-Wallis tests the distribution difference; Moran's I tests whether the under-served districts cluster <strong>in space</strong>. If they do, it's not random noise - the coverage-need mismatch has geographic structure. We build queen-contiguity weights on district governorate centroids (proxy for admin-level boundaries we don't have polygons for) and run both global Moran's I (999 permutations) and Local Moran's I (LISA) classification (HH/LL/HL/LH per district).</p>
</div>

</div>


```python
# Build GeoDataFrame from feat centroids
centroids = gpd.GeoDataFrame(d[['name','stations_per_100k','cx','cy']],
    geometry=gpd.points_from_xy(d.cx, d.cy), crs='EPSG:4326').to_crs('EPSG:32636')
centroids = centroids.reset_index(drop=True)

w = KNN.from_dataframe(centroids, k=5)
w.transform = 'r'

# Global Moran's I on stations_per_100k
moran = Moran(centroids['stations_per_100k'].values, w, permutations=999)
print(f"  Global Moran's I on stations/100k:")
print(f"    I = {moran.I:.3f}   E(I) = {moran.EI:.3f}")
print(f"    z = {moran.z_sim:.3f}   p (1-tailed) = {moran.p_sim:.4f}")

# Local Moran (LISA)
lisa = Moran_Local(centroids['stations_per_100k'].values, w, permutations=999, seed=42)
centroids['local_I'] = lisa.Is
centroids['lisa_q']  = lisa.q  # 1=HH 2=LH 3=LL 4=HL
centroids['lisa_sig']= lisa.p_sim < 0.05
q_map = {1:'HH',2:'LH',3:'LL',4:'HL'}
centroids['lisa_label'] = centroids['lisa_q'].map(q_map).where(centroids['lisa_sig'], 'ns')
print(f"\n  Significant LISA clusters (p<0.05):")
print(centroids[centroids['lisa_sig']].groupby('lisa_label').size().to_dict())
```

 <p style="margin:8px 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#E9C46A;">Test:</strong> Mann-Whitney U on the distribution of catchment population (LRT vs post-2012 metro)</p>
 <p style="margin:8px 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#E86F51;">Effect size:</strong> Cliff's δ</p>

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">For each LRT and each post-2012 metro station: spatial join to Phase 1 population hexes within 2 km, sum Population_2018.</p>

```python
# H2 · 2-km catchment population for each LRT station vs each post-2012 metro station.
post2012_metro = metro_pts[(metro_pts.line == 'Line 3') & (metro_pts.opening_year >= 2014)].copy()
lrt_pts_h2 = gpd.GeoDataFrame(lrt.dropna(subset=['lat','lon']),
    geometry=gpd.points_from_xy(lrt.dropna(subset=['lat','lon']).lon, lrt.dropna(subset=['lat','lon']).lat),
    crs='EPSG:4326').to_crs('EPSG:32636')

def catchment_pop(stations_gdf, hexes_gdf, radius_m=2000):
    results = []
    for _, r in stations_gdf.iterrows():
        buf = r.geometry.buffer(radius_m)
        hit = hexes_gdf[hexes_gdf.geometry.within(buf) | hexes_gdf.geometry.intersects(buf)]
        results.append(hit['Population_2018'].sum())
    return np.array(results)

lrt_catch   = catchment_pop(lrt_pts_h2, p1_pop_gdf)
metro_catch = catchment_pop(post2012_metro, p1_pop_gdf)

u_stat, p = stats.mannwhitneyu(lrt_catch, metro_catch, alternative='less')
delta = cliffs_delta(lrt_catch, metro_catch)
show_metrics({
    'LRT stations (n)':              len(lrt_catch),
    'Metro L3 post-2012 (n)':        len(metro_catch),
    'LRT median 2-km catchment':     f'{int(np.median(lrt_catch)):,}' if len(lrt_catch) else 'n/a',
    'Metro median 2-km catchment':   f'{int(np.median(metro_catch)):,}' if len(metro_catch) else 'n/a',
    'Mann-Whitney U (LRT < metro)':  f'U={u_stat:.0f}, p={p:.4f}',
    "Cliff's δ (effect size)":       round(delta, 3),
}, title='H2 · LRT catchment deficit', highlight={"Cliff's δ (effect size)"})
```

 <p style="margin:8px 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#58A6FF;">H₀:</strong> The Ring Road BRT corridor carries the same pre-BRT informal-transit demand density as an equivalent-length non-Ring-Road corridor.</p>

 <p style="margin:8px 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#E9C46A;">Test:</strong> Mann-Whitney U on per-segment informal daily boardings in a 500 m buffer · matched control corridor sampled by distance to CBD</p>
 <p style="margin:8px 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong style="color:#E86F51;">Effect size:</strong> Cliff's δ</p>

<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">For each BRT station's 500 m buffer, sum informal boardings. Construct a matched non-Ring-Road control by sampling random points from the urbanized area at similar distance-to-center. Compare distributions.</p>

```python
# H3 · BRT corridor informal demand vs matched non-corridor controls.
from shapely.geometry import Point as _Pt
brt_pts_h3 = gpd.GeoDataFrame(brt.dropna(subset=['lat','lon']),
    geometry=gpd.points_from_xy(brt.dropna(subset=['lat','lon']).lon, brt.dropna(subset=['lat','lon']).lat),
    crs='EPSG:4326').to_crs('EPSG:32636')

if len(brt_pts_h3):
    brt_buf2 = brt_pts_h3.copy(); brt_buf2['geometry'] = brt_buf2.buffer(500)
    q4_gdf_h3 = q4_gdf[q4_gdf.Informal_Share_Daily.notna()].copy()
    brt_informal = []
    for _, r in brt_buf2.iterrows():
        hits = q4_gdf_h3[q4_gdf_h3.geometry.within(r.geometry)]
        brt_informal.append(hits['Daily_Informal_BA'].sum())
    brt_informal = np.array(brt_informal)

    rng = np.random.default_rng(7)
    minx, miny, maxx, maxy = q4_gdf_h3.total_bounds
    brt_buf_union = brt_buf2.geometry.union_all() if hasattr(brt_buf2.geometry,'union_all') else brt_buf2.geometry.unary_union
    controls = []; attempts = 0
    while len(controls) < len(brt_pts_h3) and attempts < 2000:
        x, y = rng.uniform(minx, maxx), rng.uniform(miny, maxy)
        pt = _Pt(x, y)
        if not pt.within(brt_buf_union):
            hits = q4_gdf_h3[q4_gdf_h3.geometry.within(pt.buffer(500))]
            controls.append(hits['Daily_Informal_BA'].sum())
        attempts += 1
    controls = np.array(controls)

    u_stat, p = stats.mannwhitneyu(brt_informal, controls, alternative='greater')
    delta = cliffs_delta(brt_informal, controls)
    show_metrics({
        'BRT corridor median informal boardings': f'{int(np.median(brt_informal)):,}',
        'Matched control median':                 f'{int(np.median(controls)):,}',
        'Mann-Whitney U (BRT > control)':         f'U={u_stat:.0f}, p={p:.4f}',
        "Cliff's δ (effect size)":                round(delta, 3),
    }, title='H3 · BRT corridor match', highlight={"Cliff's δ (effect size)"})
else:
    brt_informal = np.array([]); controls = np.array([])
    print('Skipped H3: no BRT coords')
```
