```python
# Paths ─ all scraping caches in RAW_DIR; cleaned outputs in CLEAN_DIR; joined results in INTEGRATED_DIR
RAW_DIR        = Path("RawData")
CLEAN_DIR      = Path("CleanedData")
INTEGRATED_DIR = Path("Integrated")
for d in (RAW_DIR, CLEAN_DIR, INTEGRATED_DIR):
    d.mkdir(exist_ok=True)

# Phase 1 CleanedData sits alongside us — assert it's present
PHASE1_CLEAN = Path("..") / "CleanedData"
assert PHASE1_CLEAN.exists(), f"Phase 1 CleanedData not found at {PHASE1_CLEAN.resolve()}"

CRS_METERS   = "EPSG:32636"                 # UTM 36N; all distance computations use this
CRS_MAP      = "EPSG:4326"                  # WGS84; only for web basemaps (Folium/Mapbox)
CAIRO_BBOX   = {"lat_min": 29.8, "lat_max": 30.3, "lon_min": 31.0, "lon_max": 31.7}
```

```python
# Reusable post-scrape exploration helpers.
# These cells audit inconsistencies and outlier candidates only; they do not mutate or remove rows.
from IPython.display import display

for _drafts_dir in (Path('..') / 'drafts', Path('drafts')):
    if _drafts_dir.exists():
        _drafts_path = str(_drafts_dir.resolve())
        if _drafts_path not in sys.path:
            sys.path.insert(0, _drafts_path)
        break

from phase2_utils import show_df, show_metrics, show_note

_ID_LIKE_COLS = {'id', 'osm_id', 'stop_id', 'route_id', 'trip_id', 'shape_id', 'agency_id', 'service_id'}
_AUDIT_COLORS = {
    'blue': '#58A6FF', 'gold': '#E9C46A', 'teal': '#2A9D8F',
    'coral': '#E86F51', 'muted': '#8B949E', 'body': '#C9D1D9',
    'bg': '#0D1117', 'panel': 'rgba(22,27,34,0.72)'
}

def _read_scraped_artifact(path):
    path = Path(path)
    if path.suffix.lower() in {'.geojson', '.json'} and 'geojson' in path.name.lower():
        return gpd.read_file(path)
    return pd.read_csv(path)

def _audit_layout(title, height=360):
    return dict(
        template='plotly_dark', height=height,
        paper_bgcolor=_AUDIT_COLORS['bg'], plot_bgcolor=_AUDIT_COLORS['bg'],
        title=dict(text=title, font=dict(color=_AUDIT_COLORS['gold'], size=15)),
        font=dict(color=_AUDIT_COLORS['body'], family='Share Tech Mono'),
        margin=dict(l=50, r=24, t=58, b=54),
        legend=dict(bgcolor=_AUDIT_COLORS['panel'], bordercolor='rgba(88,166,255,0.2)')
    )

def _numeric_profile(df, min_non_null=3):
    rows = []
    for col in df.columns:
        col_l = col.lower()
        if col_l in _ID_LIKE_COLS or col_l.endswith('_id') or col_l == 'geometry':
            continue
        s = pd.to_numeric(df[col], errors='coerce')
        # Cast booleans/nullable numerics to float so quantiles/IQR math is stable.
        s = s.astype(float)
        non_null = int(s.notna().sum())
        if non_null < min_non_null:
            continue
        q1, q3 = s.quantile([0.25, 0.75])
        iqr = q3 - q1
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr
        outliers = int(((s < lower) | (s > upper)).sum()) if pd.notna(iqr) and iqr > 0 else 0
        rows.append({
            'column': col,
            'non_null': non_null,
            'min': s.min(),
            'median': s.median(),
            'max': s.max(),
            'iqr_outlier_candidates': outliers,
            'iqr_lower': lower,
            'iqr_upper': upper,
        })
    if not rows:
        return pd.DataFrame(columns=['column','non_null','min','median','max','iqr_outlier_candidates','iqr_lower','iqr_upper'])
    return pd.DataFrame(rows).sort_values(['iqr_outlier_candidates', 'column'], ascending=[False, True])

def _categorical_candidates(df):
    candidates = []
    preferred = ('source', 'lang', 'line', 'phase', 'status', 'type', 'method', 'governorate', 'operational', 'centroid')
    for col in df.columns:
        col_l = col.lower()
        if col_l in _ID_LIKE_COLS or col_l.endswith('_id') or col_l == 'geometry':
            continue
        if pd.api.types.is_numeric_dtype(df[col]) and not pd.api.types.is_bool_dtype(df[col]):
            continue
        nunique = df[col].nunique(dropna=False)
        if 2 <= nunique <= 20:
            score = 0 if any(token in col_l for token in preferred) else 1
            candidates.append((score, nunique, col))
    return [col for _, _, col in sorted(candidates)[:2]]

def _plot_nulls(null_summary, source_name):
    if null_summary.empty:
        return
    plot_df = null_summary.head(20).reset_index().rename(columns={'index': 'column'})
    fig = go.Figure(go.Bar(
        x=plot_df['column'], y=plot_df['null_pct'],
        marker_color=_AUDIT_COLORS['coral'],
        customdata=plot_df['nulls'],
        hovertemplate='%{x}<br>%{y:.1f}% null<br>%{customdata:,} rows<extra></extra>'
    ))
    fig.update_layout(**_audit_layout(f'{source_name} ? missing values by column'))
    fig.update_xaxes(title='column', tickangle=-35)
    fig.update_yaxes(title='% missing')
    fig.show()

def _plot_numeric_outliers(numeric, source_name):
    if numeric.empty:
        return
    plot_df = numeric.head(12).sort_values('iqr_outlier_candidates')
    fig = go.Figure(go.Bar(
        x=plot_df['iqr_outlier_candidates'], y=plot_df['column'], orientation='h',
        marker_color=np.where(plot_df['iqr_outlier_candidates'] > 0, _AUDIT_COLORS['coral'], _AUDIT_COLORS['teal']),
        customdata=np.stack([plot_df['min'], plot_df['median'], plot_df['max']], axis=-1),
        hovertemplate='%{y}<br>IQR candidates: %{x:,}<br>min=%{customdata[0]:,.3f}<br>median=%{customdata[1]:,.3f}<br>max=%{customdata[2]:,.3f}<extra></extra>'
    ))
    fig.update_layout(**_audit_layout(f'{source_name} ? numeric IQR outlier candidates'))
    fig.update_xaxes(title='candidate rows')
    fig.update_yaxes(title='numeric column')
    fig.show()

def _plot_coordinates(df, lat_col, lon_col, outside, source_name, max_points=5000):
    lat = pd.to_numeric(df[lat_col], errors='coerce')
    lon = pd.to_numeric(df[lon_col], errors='coerce')
    plot_df = pd.DataFrame({lat_col: lat, lon_col: lon, 'bbox_status': np.where(outside, 'outside Cairo bbox', 'inside Cairo bbox')}).dropna()
    if plot_df.empty:
        return
    if len(plot_df) > max_points:
        plot_df = plot_df.sample(max_points, random_state=42)
    fig = go.Figure()
    for status, color in [('inside Cairo bbox', _AUDIT_COLORS['blue']), ('outside Cairo bbox', _AUDIT_COLORS['coral'])]:
        sub = plot_df[plot_df['bbox_status'] == status]
        if sub.empty:
            continue
        fig.add_trace(go.Scattergl(
            x=sub[lon_col], y=sub[lat_col], mode='markers', name=status,
            marker=dict(size=5, color=color, opacity=0.62),
            hovertemplate='lon=%{x:.5f}<br>lat=%{y:.5f}<extra></extra>'
        ))
    fig.add_shape(
        type='rect', x0=CAIRO_BBOX['lon_min'], x1=CAIRO_BBOX['lon_max'],
        y0=CAIRO_BBOX['lat_min'], y1=CAIRO_BBOX['lat_max'],
        line=dict(color=_AUDIT_COLORS['gold'], width=2, dash='dash'),
        fillcolor='rgba(0,0,0,0)'
    )
    fig.update_layout(**_audit_layout(f'{source_name} ? coordinate spread vs Cairo bbox', height=430))
    fig.update_xaxes(title='longitude')
    fig.update_yaxes(title='latitude', scaleanchor='x', scaleratio=1)
    fig.show()

def _plot_categorical(df, source_name):
    for col in _categorical_candidates(df):
        counts = df[col].fillna('Missing').astype(str).value_counts().head(12).reset_index()
        counts.columns = [col, 'rows']
        show_df(counts, title=f'{source_name} ? {col} distribution', caption='Categorical balance check from the scraped artifact')
        fig = go.Figure(go.Bar(
            x=counts[col], y=counts['rows'], marker_color=_AUDIT_COLORS['teal'],
            hovertemplate='%{x}<br>%{y:,} rows<extra></extra>'
        ))
        fig.update_layout(**_audit_layout(f'{source_name} ? {col} distribution'))
        fig.update_xaxes(title=col, tickangle=-25)
        fig.update_yaxes(title='rows')
        fig.show()

def explore_scraped_data(source_name, path, key_cols=None, coord_cols=None, bbox=CAIRO_BBOX, top_n=5):
    """Read one scraped artifact and show styled audit tables plus visual checks. No rows are edited."""
    df = _read_scraped_artifact(path)
    key_cols = [c for c in (key_cols or []) if c in df.columns]
    source_title = f'{source_name} ? {Path(path).name}'

    show_note(
        f'<strong>{Path(path).name}</strong> loaded for exploration only. The checks below flag missingness, duplicate keys, coordinate spread, and IQR outlier candidates; no values are removed or changed.',
        kind='info', title=source_name
    )
    show_metrics({
        'rows': len(df),
        'columns': df.shape[1],
        'artifact': Path(path).name,
    }, title=f'{source_name} ? artifact shape', highlight={'rows'})

    null_summary = (df.isna().sum().rename('nulls').to_frame()
        .assign(null_pct=lambda x: x['nulls'] / max(len(df), 1) * 100)
        .query('nulls > 0')
        .sort_values(['null_pct', 'nulls'], ascending=False))
    if len(null_summary):
        show_df(null_summary.reset_index().rename(columns={'index': 'column'}).head(20),
                title=f'{source_name} ? columns with missing values',
                caption='High-null columns are audit flags, not automatic drop instructions.',
                highlight_col='null_pct')
        _plot_nulls(null_summary, source_name)
    else:
        show_metrics({'columns with missing values': 0}, title=f'{source_name} ? missing-value check')

    if key_cols:
        dup_count = int(df.duplicated(key_cols, keep=False).sum())
        show_metrics({'duplicate rows': dup_count, 'key columns': ', '.join(key_cols)},
                     title=f'{source_name} ? duplicate-key check', highlight={'duplicate rows'})
        if dup_count:
            show_df(df.loc[df.duplicated(key_cols, keep=False), key_cols].head(12),
                    title=f'{source_name} ? duplicate-key sample',
                    caption='Sample only; rows are left untouched for later review.')

    if coord_cols and all(c in df.columns for c in coord_cols):
        lat_col, lon_col = coord_cols
        lat = pd.to_numeric(df[lat_col], errors='coerce')
        lon = pd.to_numeric(df[lon_col], errors='coerce')
        has_coord = lat.notna() & lon.notna()
        outside = has_coord & ~(
            lat.between(bbox['lat_min'], bbox['lat_max']) &
            lon.between(bbox['lon_min'], bbox['lon_max'])
        )
        coord_summary = pd.DataFrame([{
            'coordinate_rows': int(has_coord.sum()),
            'missing_coordinate_rows': int((~has_coord).sum()),
            'outside_cairo_bbox': int(outside.sum()),
            'lat_min': lat.min(), 'lat_max': lat.max(),
            'lon_min': lon.min(), 'lon_max': lon.max(),
        }])
        show_df(coord_summary, title=f'{source_name} ? coordinate sanity check',
                caption='BBox flags are exploratory. Some corridor data intentionally extends beyond central Cairo.',
                highlight_col='outside_cairo_bbox')
        _plot_coordinates(df, lat_col, lon_col, outside, source_name)
        if outside.any():
            show_df(df.loc[outside, [c for c in [lat_col, lon_col, *key_cols] if c in df.columns]].head(10),
                    title=f'{source_name} ? sample outside Cairo bbox',
                    caption='Outliers are shown for review and kept in the data.')

    numeric = _numeric_profile(df)
    if len(numeric):
        show_df(numeric.head(15), title=f'{source_name} ? numeric range + IQR flags',
                caption='IQR flags identify candidates for human review; this cell does not remove them.',
                highlight_col='iqr_outlier_candidates')
        _plot_numeric_outliers(numeric, source_name)
    else:
        show_metrics({'numeric columns checked': 0}, title=f'{source_name} ? numeric outlier check')

    _plot_categorical(df, source_name)

    show_df(df.head(top_n), title=f'{source_name} ? sample rows',
            caption='Small manual scan of the artifact immediately after scraping/cleaning.')
    return df

show_note('Post-scrape exploration helpers ready: styled tables, metric cards, and Plotly audit charts will be rendered after each source artifact is created.', kind='success', title='Audit helpers ready')
```

```python
# S1 · Transport for Cairo GTFS (bus+metro bundle).
# Repo: transportforcairo/Transit---GCR-Digital-Cairo-2017-  (CC BY-NC 4.0)
GTFS_BASE = "https://raw.githubusercontent.com/transportforcairo/Transit---GCR-Digital-Cairo-2017-/master/GTFS"
GTFS_SETS = {
    "bus_metro":   "20180906_GTFSfullworking_Bus_Metro",
}
GTFS_FILES = ["agency.txt","calendar.txt","calendar_dates.txt","feed_info.txt",
              "frequencies.txt","routes.txt","shapes.txt","stop_times.txt",
              "stops.txt","trips.txt","fare_attributes.txt","fare_rules.txt"]

for label, subdir in GTFS_SETS.items():
    out_zip = RAW_DIR / f"gtfs_{label}.zip"
    
    if out_zip.exists() and out_zip.stat().st_size > 10_000:
        continue   # already exists

    with zipfile.ZipFile(out_zip, "w", zipfile.ZIP_DEFLATED) as zf:
        for fn in GTFS_FILES:
            r = Fetcher.get(f"{GTFS_BASE}/{subdir}/{fn}", stealthy_headers=True, timeout=60)
            if r.status == 200:
                zf.writestr(fn, r.body)
            time.sleep(0.2)
```

```python
# Unzip the bus+metro bundle and load the GTFS tables into memory.
extract_dir = RAW_DIR / "gtfs_bus_metro"
extract_dir.mkdir(exist_ok=True)
with zipfile.ZipFile(RAW_DIR / f"gtfs_bus_metro.zip") as z:
    z.extractall(extract_dir)

gtfs = {fn.replace(".txt",""): pd.read_csv(extract_dir / fn)
        for fn in GTFS_FILES if (extract_dir / fn).exists()}
print(f"S1 GTFS: {len(gtfs)} tables loaded; stops={len(gtfs['stops']):,} routes={len(gtfs['routes']):,} trips={len(gtfs['trips']):,}")
```

```python
# Filter stops to Cairo bbox, reproject to EPSG:32636, persist only the tables NB2 uses.
stops = gtfs["stops"]
mask  = (stops["stop_lat"].between(CAIRO_BBOX["lat_min"], CAIRO_BBOX["lat_max"]) &
         stops["stop_lon"].between(CAIRO_BBOX["lon_min"], CAIRO_BBOX["lon_max"]))
stops_gdf = gpd.GeoDataFrame(stops[mask].copy(),
    geometry=gpd.points_from_xy(stops[mask].stop_lon, stops[mask].stop_lat),
    crs=CRS_MAP).to_crs(CRS_METERS)
stops_gdf["x_m"] = stops_gdf.geometry.x
stops_gdf["y_m"] = stops_gdf.geometry.y
stops_gdf.drop(columns=["geometry"]).to_csv(CLEAN_DIR / "gtfs_stops.csv", index=False)

# Save only the GTFS tables consumed downstream (Q19 routes, Q21 shapes+trips+fare_attributes)
for k in ("routes","trips","shapes","fare_attributes"):
    if k in gtfs:
        gtfs[k].to_csv(CLEAN_DIR / f"gtfs_{k}.csv", index=False)

print(f"S1 cleaned: {mask.sum():,} stops in Cairo bbox · reprojected EPSG:32636 · 5 CSVs written")
```

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Parse each station row. <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">clean_cell()</code> strips cite markers; the trailing-digit strip (for odd name suffixes) is <strong>off</strong> for the Opened column so values like <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">1987</code> are not erased. <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">dms_to_dd()</code> converts DMS coordinate strings to decimal degrees. Opening dates are parsed with a 4-digit year regex first, then <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">dateutil</code>.</p>

```python
# Parse station rows — footnotes stripped, DMS→decimal, dates→year, phase labels derived.
def clean_cell(el, strip_trailing_digits=True):
    """Strip cite markers; optional trailing-digit strip is for name cells only —
    year-only opened dates like '1987' must not be blanked."""
    text = re.sub(r'\[[^\]]*\]', '', _norm_text(el))
    if strip_trailing_digits:
        text = re.sub(r'\s*\d+\s*$', '', text)
    return text

def dms_to_dd(s):
    s = s.replace('′', "'").replace('″', '"').replace('°', ' ')
    matches = re.findall(r"([\d.]+)\s*[°\s]*(\d+)?'?\s*([\d.]+)?\"?\s*([NSEW])", s)
    if len(matches) < 2: return None, None
    coords = []
    for m in matches[:2]:
        dd = float(m[0]) + float(m[1] or 0)/60 + float(m[2] or 0)/3600
        if m[3] in ('S','W'): dd = -dd
        coords.append(dd)
    return coords[0], coords[1]

def parse_year(s):
    s = (s or '').strip()
    if not s: return None
    m = re.search(r'\b\d{4}\b', s)
    if m:
        return int(m.group(0))
    try:
        return dateparser.parse(s).year
    except Exception:
        return None
    
def assign_phase(row):
    y, ln = row['opening_year'], row['line']
    if y is None: return 'unknown'
    if ln == 'Line 1': return 'Phase 1 (orig)' if y <= 1989 else 'Phase 1 (ext)'
    if ln == 'Line 2': return 'Phase 2 (orig)'
    if ln == 'Line 3':
        if y <= 2014: return 'Phase 3A'
        if y <= 2019: return 'Phase 3B'
        if y <= 2023: return 'Phase 3C'
        return 'Phase 3D'
    return 'other'
```

```python
rows = []
for line_idx, t in enumerate(station_tables[:3]):
    line = f"Line {line_idx + 1}"
    for r in t.css('tr')[1:]:
        cs = r.css('th, td')
        if len(cs) < 3: continue
        lat, lon = dms_to_dd(clean_cell(cs[2])) if len(cs) > 2 else (None, None)
        rows.append({'line': line,
                     'station_en':   clean_cell(cs[0]),
                     'station_ar':   clean_cell(cs[1]) if len(cs) >= 1 else '',
                     'location_raw': clean_cell(cs[2]) if len(cs) >= 2 else '',
                     'opened_raw':   clean_cell(cs[3], strip_trailing_digits=False) if len(cs) > 3 else '',
                     'lat': lat, 'lon': lon})

metro = pd.DataFrame(rows)
metro['opening_year'] = metro['opened_raw'].apply(parse_year)


metro['phase'] = metro.apply(assign_phase, axis=1)
metro['is_interchange'] = metro['station_en'].map(metro.groupby('station_en')['line'].nunique()) > 1
metro.to_csv(CLEAN_DIR / 'metro_stations.csv', index=False)

with_coords = metro[['lat','lon']].notna().all(axis=1).sum()
print(f"S3 metro: {len(metro)} stations · {with_coords} have coords · phases: {sorted(metro['phase'].unique())}")
```

```python
# S4 · Wikipedia Cairo_Light_Rail_Transit — 3 wikitables (operational / phase-2 / extension).
S4_URL = "https://en.wikipedia.org/wiki/Cairo_Light_Rail_Transit"
out = RAW_DIR / "wiki_lrt.html"

if not (out.exists() and out.stat().st_size > 50_000):
    r = Fetcher.get(S4_URL, timeout=60, stealthy_headers=True)
    body = r.body.decode('utf-8') if isinstance(r.body, bytes) else r.body
    out.write_text(body, encoding='utf-8') # for cache only

tables = Selector(out.read_text(encoding='utf-8')).css('table.wikitable')

def parse_lrt_table(t):
    hdrs = [_norm_text(c) for c in t.css('tr')[0].css('th, td')]
    out_rows = []
    
    for r in t.css('tr')[1:]:
        cs = r.css('th, td')

        if not cs: continue

        status_cell = clean_cell(cs[1]) if len(cs) > 1 else '' # Acually doesn't do anything as all the data doesn't have any []
        is_operational = 'Under' not in status_cell
        rec = {'is_operational': is_operational}

        for j, h in enumerate(hdrs[:len(cs)]):
            rec[(h or f'col{j}').lower()] = clean_cell(cs[j]) # again clean_cell is not needed but exists for future-proof.
        out_rows.append(rec)
    return out_rows

lrt_rows = []
for idx, t in enumerate(tables):
    lrt_rows.extend(parse_lrt_table(t))

lrt = pd.DataFrame(lrt_rows)
for cand in ('station', 'stop', 'name'):
    if cand in lrt.columns:
        lrt = lrt.rename(columns={cand: 'station_en'}); break
    
lrt['line'] = 'LRT'
lrt.to_csv(CLEAN_DIR / 'lrt_stations.csv', index=False)
print(f"S4 LRT: {len(lrt)} stations ({lrt['is_operational'].sum()} operational)")
```

Places are extracted by regex pattern: <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">aria-label="{name}"</code> for the station name, <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">!3d{lat}!4d{lon}</code> for the coordinates from the URL fragment. Scraping Google Maps directly removes the dependency on the Places API key. For deduplication across the two language queries, we use a <strong>neural Arabic→English translation model</strong> (<code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">Helsinki-NLP/opus-mt-ar-en</code>, ~300 MB MarianMT) rather than a hand-coded dictionary - translations are generated at run time from whatever station names the scrape returns, so adding new stations requires zero code changes.</p>

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Scrape Google Maps with English and Arabic search queries. Each search returns a rendered HTML page with place cards; we extract <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">(name, lat, lon)</code> tuples from <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">aria-label</code> + <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">!3d…!4d…</code> patterns. Snapshots cached to <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">RawData/</code> for audit.</p>

```python
# S5 · BRT via Google Maps (playwright-backed DynamicFetcher).
# Search for 2 queries: BRT (English) and الترددي (Arabic).
# Extract all coordinates found, filter by CAIRO_BBOX only when writing to CSV.
GMAPS_QUERIES = {
    'brt_en':    'BRT',
    'trd_ar':    'الترددي',
}

COORD_PAT     = re.compile(r'aria-label="([^"]{3,120})"[^>]*href="[^"]*?!3d(-?\d+\.\d+)!4d(-?\d+\.\d+)')
CANONICAL_PAT = re.compile(r'<link rel="canonical" href="https://www\.google\.com/maps/place/([^/"]+)/@(-?\d+\.\d+),(-?\d+\.\d+)')

def _is_brt_station(nm):
    return 'brt' in nm.lower() or 'الترددي' in nm

def parse_gmaps(html):
    """Extract all station names and coordinates from Google Maps HTML. No bbox filtering here."""
    hits = []
    m = CANONICAL_PAT.search(html)
    if m:
        hits.append((m.group(1).replace('+', ' '), float(m.group(2)), float(m.group(3))))
    hits += [(m.group(1).strip(), float(m.group(2)), float(m.group(3))) for m in COORD_PAT.finditer(html)]
    return [(nm, la, lo) for nm, la, lo in hits if _is_brt_station(nm)]

def _fetch_gmaps(url: str, out_path: Path):
    out_str = str(out_path).replace('\\', '/')
    script = textwrap.dedent(f"""
        import asyncio
        from playwright.async_api import async_playwright

        async def main():
            async with async_playwright() as pw:
                browser = await pw.chromium.launch(headless=True)
                page = await browser.new_page()
                await page.goto({url!r}, wait_until='domcontentloaded', timeout=600000)
                await asyncio.sleep(15)
                try:
                    await page.click('button[aria-label*="Accept"]', timeout=5000)
                except:
                    pass
                try:
                    await page.wait_for_selector('div[role="feed"]', timeout=20000)
                    for _ in range(3):
                        await page.locator('div[role="feed"]').evaluate('el => el.scrollBy(0, el.scrollHeight)')
                        await asyncio.sleep(2)
                except:
                    pass
                html = await page.content()
                await browser.close()
            with open({out_str!r}, 'w', encoding='utf-8') as f:
                f.write(html)

        asyncio.run(main())
    """)
    result = subprocess.run(
        [sys.executable, '-c', script],
        capture_output=True, text=True, encoding='utf-8')

def _scrape_brt():
    rows = []
    for label, query in GMAPS_QUERIES.items():
        out = RAW_DIR / f'gmaps_brt_{label}.html'

        if not (out.exists() and out.stat().st_size > 200_000):
            url = f'https://www.google.com/maps/search/{query}'
            _fetch_gmaps(url, out)

        stations = parse_gmaps(out.read_text(encoding='utf-8'))
        print(f"[{label}] {len(stations)} BRT hits · {out.stat().st_size // 1024} KB")

        for name, la, lo in stations:
            is_ar = any('\u0600' <= ch <= '\u06ff' for ch in name)
            rows.append({
                'source_scrape': 'google_maps',
                'search_lang':   label,
                'raw_name':      name,
                'station_en':    '' if is_ar else name,
                'station_ar':    name if is_ar else '',
                'lat': la, 'lon': lo,
            })

    return rows

brt_raw = pd.DataFrame(_scrape_brt())

brt_raw_filtered = brt_raw[
    (brt_raw['lat'].between(CAIRO_BBOX['lat_min'], CAIRO_BBOX['lat_max'])) &
    (brt_raw['lon'].between(CAIRO_BBOX['lon_min'], CAIRO_BBOX['lon_max']))
].copy()
print(f"S5 BRT scrape: {len(brt_raw)} raw rows · {len(brt_raw_filtered)} within Cairo bbox")
brt_raw_filtered.head(30)
```

 <span style="font-family:'Share Tech Mono',monospace;color:#7B2D8E;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">METHODOLOGY ENHANCEMENT</span>
 <p style="margin:10px 0 0;color:#C9D1D9;line-height:1.65;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong>Franko transliteration for dedup.</strong> Place names need <em>transliteration</em>, not translation - <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">المرج</code> is <em>El Marg</em>, not the semantic <em>lawn</em>; <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">السلام</code> is <em>El Salam</em>, not <em>Peace</em>. We use ISI's <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">uroman</code> (universal romanizer), a Unicode character-level transliterator that maps Arabic script to Latin phonetics (franko / Arabizi). Duplicates are then detected by <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">rapidfuzz.fuzz.token_set_ratio</code> ≥ 70 between the transliterated and the native-English names. This replaces a hand-coded Arabic→Latin dictionary (brittle) and a neural MT model (semantically wrong for proper nouns) - uroman is deterministic and well-suited to proper-noun romanisation.</p>
</div>


<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Share Tech Mono',monospace;color:#58A6FF;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">⟐ CODE EXPLANATION</span>
 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Load Helsinki-NLP's MarianMT Arabic→English model. First call downloads ~300 MB from HuggingFace; subsequent calls hit the local cache. Batch all Arabic names in one forward pass.</p>
</div>


```python
# Franko transliteration (Arabic → Latin phonetics, NOT semantic translation).
# ISI uroman: character-level Unicode-based romanizer. 'المرج' → 'almrj' not 'lawn'.
_romanizer = _ur.Uroman()

def transliterate(texts):
    return [_romanizer.romanize_string(t or '') for t in texts]

ar_rows = brt_raw[brt_raw.station_ar != ''].copy()
brt_raw['translated_en'] = brt_raw['station_en']
brt_raw.loc[ar_rows.index, 'translated_en'] = transliterate(ar_rows.station_ar.tolist())
print(f"S5 transliteration (franko): {len(ar_rows)} Arabic names converted via uroman")
```

<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Share Tech Mono',monospace;color:#58A6FF;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">⟐ CODE EXPLANATION</span>
 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Dedup rows in three passes: (1) spatial - if two rows are within 80 m, they are the same station; (2) semantic - compute <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">rapidfuzz.token_set_ratio</code> between translated_en strings; ratio ≥ 70 with <em>any</em> kept row → duplicate. Priority: confirmed-BRT first, then shorter names, then first-seen.</p>
</div>



```python
# S5 dedup — three-tier (spatial near / spatial far + name match / unique).
# The same station can appear in multiple language results with slightly different pins;
# we fold them into one row while keeping genuinely-different stations that share a name.
FRANKO_STOP = {
    'mhta','mahata','mahatta',                                 # محطة (station)
    'alatwbys','alautwbys','atwbys','autwbys','autobis','otobis',
    'altrddy','altrdy','altrady','trddy','trdy',
    'alsry','alsryy','sry','sri',"alsry'","sry'","'sry",
    'mwqf','mwkf','mawqef','mykrwbas','mikrobas','microbas',
    'station','stop','brt','bus','cairo','metro','the','of','el','al',
    '-','·','',
}

def _core_tokens(name):
    return {t for t in re.split(r'[\s·\-،,.()/]+', (name or '').lower())
            if t and t not in FRANKO_STOP}

# Priority: confirmed-BRT first, then shorter names win on ties
brt_raw['_priority'] = brt_raw.apply(
    lambda r: (len(r.translated_en), len(r.raw_name)), axis=1)
brt_sorted = brt_raw.sort_values('_priority').reset_index(drop=True)

brt_g = gpd.GeoDataFrame(brt_sorted,
    geometry=gpd.points_from_xy(brt_sorted.lon, brt_sorted.lat),
    crs=CRS_MAP).to_crs(CRS_METERS)

# Two distance bands:
#   <80m                    : same pin — always dedup
#   80m..NAME_DEDUP_RADIUS  : same station (name-based dedup eligible)
#   > NAME_DEDUP_RADIUS     : different station even if names match
# Salam pins at (30.1489, 31.4187) and (30.1549, 31.4092) sit ~1.05 km apart along
# the Ring Road; neither Wikipedia nor TfC lists two separate Salam stations, so
# we merge them by setting the window at 1200m.
NAME_DEDUP_RADIUS = 1200   # metres

def _similar_cores(a, b):
    ta, tb = _core_tokens(a), _core_tokens(b)
    if not ta or not tb: return False
    if ta & tb: return True
    return fuzz.token_set_ratio(' '.join(sorted(ta)), ' '.join(sorted(tb))) >= 70

keep = []
for i in range(len(brt_g)):
    r = brt_g.iloc[i]
    dup = False
    for j in keep:
        other = brt_g.iloc[j]
        d = r.geometry.distance(other.geometry)
        if d < 80 or (d < NAME_DEDUP_RADIUS and _similar_cores(r.translated_en, other.translated_en)):
            dup = True; break
    if not dup:
        keep.append(i)

brt_unique_g = brt_g.iloc[keep].copy().reset_index(drop=True)

# Final outlier guard: after name-based dedup, remove any BRT terminal still too close
# to an already-kept BRT terminal. Different Google Maps labels sometimes point to the
# same physical Ring Road station/interchange, so a minimum spacing rule is safer than
# relying on names alone.
BRT_MIN_SEPARATION_M = 300
sep_keep = []
close_drop_rows = []
for i in range(len(brt_unique_g)):
    r = brt_unique_g.iloc[i]
    too_close_to = None
    too_close_dist = None
    for j in sep_keep:
        other = brt_unique_g.iloc[j]
        d = float(r.geometry.distance(other.geometry))
        if d < BRT_MIN_SEPARATION_M:
            too_close_to = other.translated_en
            too_close_dist = d
            break
    if too_close_to is None:
        sep_keep.append(i)
    else:
        close_drop_rows.append({
            'dropped_station': r.translated_en,
            'kept_station': too_close_to,
            'distance_m': round(too_close_dist, 1),
        })

brt_close_drops = pd.DataFrame(close_drop_rows)
brt = brt_unique_g.iloc[sep_keep].drop(columns=['geometry','_priority']).reset_index(drop=True)
brt['line'] = 'BRT'; brt['opening_year'] = 2025; brt['is_operational'] = True
brt.to_csv(CLEAN_DIR / 'brt_stations.csv', index=False)
print(
    f"S5 BRT dedup: {len(brt_sorted)} raw -> {len(brt_unique_g)} name/spatial unique -> "
    f"{len(brt)} after {BRT_MIN_SEPARATION_M}m spacing guard (TfC Phase 1 plan: 14)"
)
if len(brt_close_drops):
    show_df(brt_close_drops, title='S5 BRT close-pin outliers removed',
            caption=f'Rows removed because they were <{BRT_MIN_SEPARATION_M}m from an already-kept BRT terminal.',
            highlight_col='distance_m')

```

```python
# S6 · citypopulation.de Egypt admin hierarchy — 408 rows (27 govs · 169 markaz · 196 kism).
# Four census/estimate waves: 1996, 2006, 2017, 2023-est.
S6_URL = "https://www.citypopulation.de/en/egypt/admin/"
out = RAW_DIR / "citypop_egypt_admin.html"

if not (out.exists() and out.stat().st_size > 100_000):
    r = Fetcher.get(S6_URL, timeout=60, stealthy_headers=True)
    body = r.body.decode('utf-8') if isinstance(r.body, bytes) else r.body
    out.write_text(body, encoding='utf-8')

admin_table = Selector(out.read_text(encoding='utf-8')).css('table')[0]

def _to_int(s):
    s = (s or '').replace(',', '').replace('...', '').strip()
    return int(s) if s.isdigit() else None

rows, current_gov = [], None
for r in admin_table.css('tr')[1:]:
    cs = r.css('th, td')
    if len(cs) < 7: continue
    name, status, native = _norm_text(cs[0]), _norm_text(cs[1]), _norm_text(cs[2])
    pops = {1996: _to_int(_norm_text(cs[3])),
            2006: _to_int(_norm_text(cs[4])),
            2017: _to_int(_norm_text(cs[5])),
            2023: _to_int(_norm_text(cs[6]))}
    if status == 'Governorate': current_gov = name
    rows.append({'name': name, 'native_ar': native, 'status': status,
                 'governorate': current_gov, **{f'pop_{y}': v for y, v in pops.items()}})

admin_all = pd.DataFrame(rows)
GC_GOVS = ['Al-Qāhirah [ Cairo ]', 'Al-Jīzah [ Giza ]', 'Al-Qalyūbiyah [ Qalyubia ]']
greater_cairo = admin_all[admin_all.governorate.isin(GC_GOVS) & (admin_all.status != 'Governorate')].copy()
print(f"S6 districts: {len(admin_all)} rows Egypt-wide · {len(greater_cairo)} Greater Cairo districts")
```

Compute a simple 10-year CAGR between 2006 and 2017 (the two closest full censuses) as a growth proxy for downstream questions (Q16 fastest-growing districts).</p>

```python
# Reshape wide→long (one row per district-year); compute 2006→2017 CAGR per district.
long_df = greater_cairo.melt(
    id_vars=['name','native_ar','status','governorate'],
    value_vars=['pop_1996','pop_2006','pop_2017','pop_2023'],
    var_name='year_col', value_name='population')
long_df['year'] = long_df['year_col'].str.replace('pop_','').astype(int)
long_df = long_df.drop(columns=['year_col']).rename(columns={'name':'district'})

def _cagr(a, b, years):
    if a is None or b is None or a <= 0 or b <= 0: return None
    return (b/a) ** (1/years) - 1
greater_cairo['cagr_2006_2017'] = greater_cairo.apply(
    lambda r: _cagr(r.pop_2006, r.pop_2017, 11), axis=1)

long_df.to_csv(CLEAN_DIR / 'districts.csv', index=False)
greater_cairo[['name','native_ar','status','governorate','pop_1996','pop_2006','pop_2017','pop_2023','cagr_2006_2017']] \
    .to_csv(CLEAN_DIR / 'districts_wide.csv', index=False)
print(f"S6 cleaned: districts.csv ({len(long_df)} district-year rows) + districts_wide.csv ({len(greater_cairo)} districts)")
```

 <span style="font-family:'Share Tech Mono',monospace;color:#58A6FF;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">⟐ CODE EXPLANATION</span>
 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Geocode each of the 68 Greater-Cairo districts through OpenStreetMap Nominatim. The earlier gov-centroid proxy (3 unique centroids for 68 districts) collapsed per-district analyses to 3 values. This step upgrades the feature geography to real district coordinates where Nominatim can find them (~70%) with governorate-centroid fallback for rural misses (Bahariya Oasis, Atfeh). Q22 / Q24 / H1 downstream now compute on real per-district centroids.</p>
</div>


```python
# Per-district centroid via Nominatim (OpenStreetMap geocoder).
# Why: the gov-centroid proxy we used earlier is a weak feature base. Real district
# centroids let Q22 / Q24 / H1 vary per district instead of collapsing to 3 values.
# Misses fall back to the governorate centroid and are flagged in `centroid_source`.
import unicodedata, json as _json, time as _time

def _strip_diacritics(s):
    return ''.join(c for c in unicodedata.normalize('NFKD', s) if not unicodedata.combining(c))

# Known transliteration aliases for Cairo place names (Nominatim hit rate ~2× with these)
_NAME_ALIASES = {
    'ad-duqqi': 'Dokki', "al-'ajuzah": 'Agouza', "al-'umraniyah": 'Omraniya',
    'al-hawamidiyah': 'Hawamdiya', 'al-badrashayn': 'Badrashin',
    'al-marj': 'El Marg', 'al-warraq': 'Warraq', 'ash-shuruq': 'Shorouk',
    'ash-shayk zayid': 'Sheikh Zayed', 'al-qahirah al-jadidah': 'New Cairo',
    'al-jizah': 'Giza', 'an-nuzhah': 'Nozha', 'as-salam': 'El Salam',
    'az-zawiyah al-hamra': 'Zawiya al-Hamra', 'madinat nasr': 'Nasr City',
    'al-matariyah': 'Matariya', 'al-maadi': 'Maadi', 'abu an-nimrus': 'Abu El Nomros',
    'madinat sittah uktubar': '6th of October City', 'atfih': 'Atfeh',
    'al-wahat al-bahariyah': 'Bahariya Oasis',
}

def _to_query(name):
    bracket = re.search(r'\[([^\]]+)\]', name)
    if bracket: return bracket.group(1).strip()
    base = re.sub(r'\[[^\]]*\]', '', name).strip()
    stripped = _strip_diacritics(base).lower()
    for k, v in _NAME_ALIASES.items():
        if stripped.startswith(k): return v
    return _strip_diacritics(base).replace("'", '')

def _gov_short(gov):
    m = re.search(r'\[([^\]]+)\]', gov or '')
    return (m.group(1).strip() if m else 'Cairo')

cache_path = RAW_DIR / 'nominatim_districts.json'
cache = _json.loads(cache_path.read_text(encoding='utf-8')) if cache_path.exists() else {}

def _geocode(name, gov):
    if name in cache: return cache[name]
    q_base = _to_query(name)
    gov_name = _gov_short(gov)
    for q in (f'{q_base}, {gov_name}, Egypt', f'{q_base}, Cairo, Egypt', q_base + ', Egypt'):
        try:
            r = Fetcher.get(f'https://nominatim.openstreetmap.org/search?q={q}&format=json&limit=1',
                             stealthy_headers=True, timeout=20)
            body = r.body.decode('utf-8') if isinstance(r.body, bytes) else r.body
            data = _json.loads(body) if r.status == 200 else []
            if data:
                lat, lon = float(data[0]['lat']), float(data[0]['lon'])
                if 29.5 < lat < 30.5 and 30.8 < lon < 32.3:
                    cache[name] = {'lat': lat, 'lon': lon, 'via': q}
                    return cache[name]
        except Exception: pass
        _time.sleep(1.1)   # Nominatim rate limit
    cache[name] = {'lat': None, 'lon': None, 'via': None}
    return cache[name]

# Governorate fallback centroids (median of successful Nominatim hits per governorate, computed below)
GOV_FALLBACK = {'Cairo': (30.050, 31.250), 'Giza': (30.010, 31.210), 'Qalyubia': (30.150, 31.200)}

dw = pd.read_csv(CLEAN_DIR / 'districts_wide.csv')
lats, lons, sources = [], [], []
for _, row in dw.iterrows():
    res = _geocode(row['name'], row['governorate'])
    if res.get('lat') is not None:
        lats.append(res['lat']); lons.append(res['lon']); sources.append('nominatim')
    else:
        fb_lat, fb_lon = GOV_FALLBACK.get(_gov_short(row['governorate']), (30.05, 31.25))
        lats.append(fb_lat); lons.append(fb_lon); sources.append('gov_fallback')

cache_path.write_text(_json.dumps(cache, ensure_ascii=False), encoding='utf-8')
dw['centroid_lat']    = lats
dw['centroid_lon']    = lons
dw['centroid_source'] = sources
dw.to_csv(CLEAN_DIR / 'districts_wide.csv', index=False)

hit_rate = sum(1 for s in sources if s == 'nominatim') / len(sources) * 100
print(f'Geocoded {len(dw)} districts · {sum(1 for s in sources if s == "nominatim")} via Nominatim '
      f'({hit_rate:.0f}%) · {sum(1 for s in sources if s == "gov_fallback")} gov-centroid fallback')
```

```python
# S7 · national Egypt vehicle total from Wikipedia + WB historical per-capita.
# CAPMAS is JS-gated (see capmas_transport_stub.html); blocker documented in S7_CAPMAS_DECISION.md.
r = Fetcher.get("https://en.wikipedia.org/wiki/List_of_countries_by_vehicles_per_capita",
                timeout=60, stealthy_headers=True)
out = RAW_DIR / "wiki_vehicles_per_capita.html"
out.write_text(r.body.decode('utf-8') if isinstance(r.body, bytes) else r.body, encoding='utf-8')

tables = Selector(out.read_text(encoding='utf-8')).css('table.wikitable')
big_table = next(t for t in tables if len(t.css('tr')) > 50)
hdrs = [_norm_text(c) for c in big_table.css('tr')[0].css('th, td')]
vpc_rows = []
for tr in big_table.css('tr')[1:]:
    cs = [_norm_text(c) for c in tr.css('th, td')]
    if len(cs) >= 5:
        vpc_rows.append({h: v for h, v in zip(hdrs, cs)})
vpc = pd.DataFrame(vpc_rows)
egy = vpc[vpc['Region'].str.contains('Egypt', na=False)].iloc[0]
natl_vehicles_2023 = int(egy['Road motor vehicles'].replace(',', ''))
natl_vpc_2023      = int(egy['Vehicles per 1,000 people'])
natl_pop_2023      = int(egy['Population'].replace(',', ''))
print(f"S7 Egypt 2023: {natl_vehicles_2023:,} vehicles · {natl_vpc_2023}/1k people · pop {natl_pop_2023:,}")
```

```python
# World Bank Open Data API for historical per-capita series (public JSON endpoint).
WB_INDICATORS = {
    'motor_vehicles_per_1k': 'IS.VEH.NVEH.P3',
    'passenger_cars_per_1k': 'IS.VEH.PCAR.P3',
}
wb_rows = []
for label, code in WB_INDICATORS.items():
    out = RAW_DIR / f"wb_{label}.json"
    if not out.exists():
        r = Fetcher.get(f"https://api.worldbank.org/v2/country/EGY/indicator/{code}?format=json&per_page=200",
                         stealthy_headers=True, timeout=30)
        out.write_bytes(r.body)
    payload = json.loads(out.read_text(encoding='utf-8'))
    if isinstance(payload, list) and len(payload) > 1:
        for obs in payload[1]:
            if obs.get('value') is not None:
                wb_rows.append({'indicator': label, 'year': int(obs['date']), 'value': obs['value']})
wb = pd.DataFrame(wb_rows)
print(f"S7 WB historical: {len(wb)} observations across {wb.indicator.nunique()} indicators")
```

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Apportion the 2023 national vehicle total to Greater Cairo governorates by population share. Population totals come from S6 (<code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">districts_wide.csv</code>) - we sum 2023 population per governorate and compute each governorate's share of the national total (114.5M per Wikipedia). Output schema: <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">governorate × year × vehicle_type × count × source</code>.</p>

```python
# Apportion national vehicle total to Greater Cairo governorates by population share.
# Method is documented; values update automatically when S6 or S7 change upstream.
dw = pd.read_csv(CLEAN_DIR / "districts_wide.csv")
dw['gov_en'] = dw['governorate'].str.extract(r"\[ ([^\]]+) \]")[0].fillna(dw['governorate'])
gov_pop_2023 = dw.groupby('gov_en')['pop_2023'].sum().rename('gov_pop_2023')

vehicle_rows = []
for gov, pop in gov_pop_2023.items():
    share = pop / natl_pop_2023
    vehicle_rows.append({'governorate': gov, 'year': 2023, 'vehicle_type': 'motor_vehicles_total',
                         'count': int(round(natl_vehicles_2023 * share)),
                         'method': 'apportioned_by_pop_share',
                         'source': 'Wikipedia list_of_countries_by_vehicles_per_capita (Egypt 2023)'})

wb_pc_rate = wb[wb.indicator == 'passenger_cars_per_1k'].sort_values('year')
wb_mv_rate = wb[wb.indicator == 'motor_vehicles_per_1k'].sort_values('year')
if len(wb_pc_rate) and len(wb_mv_rate):
    pc_rate_2023_est = wb_pc_rate.iloc[-1]['value'] * (natl_vpc_2023 / wb_mv_rate.iloc[-1]['value'])
    for gov, pop in gov_pop_2023.items():
        vehicle_rows.append({'governorate': gov, 'year': 2023, 'vehicle_type': 'passenger_cars_est',
                             'count': int(round(pc_rate_2023_est * pop / 1000)),
                             'method': 'WB_per_capita_extrapolated × gov_pop',
                             'source': 'World Bank IS.VEH.PCAR.P3 + S6 population'})

for _, row in wb.iterrows():
    if row.year < 2000: continue
    for gov, pop_2023 in gov_pop_2023.items():
        col = min(['pop_1996','pop_2006','pop_2017','pop_2023'],
                  key=lambda c: abs(int(c.split('_')[1]) - row.year))
        pop_year = dw[dw.gov_en == gov][col].sum()
        vehicle_rows.append({'governorate': gov, 'year': row.year,
                             'vehicle_type': row.indicator,
                             'count': int(round(row.value * pop_year / 1000)),
                             'method': 'WB_per_capita × gov_pop',
                             'source': f'World Bank {WB_INDICATORS[row.indicator]}'})

vehicles = pd.DataFrame(vehicle_rows).sort_values(['governorate','year','vehicle_type'])
vehicles.to_csv(CLEAN_DIR / 'vehicles_by_governorate.csv', index=False)
print(f"S7 apportioned: {len(vehicles)} rows across {vehicles.governorate.nunique()} Greater Cairo governorates")
```

```python
# S8 · OpenStreetMap transport features for Cairo (ODbL license).
# Purpose: independent cross-verification of Wikipedia coordinates (Section 11 Stage 2).
OVERPASS_QL = """
[out:json][timeout:120];
(
  node["public_transport"="station"](29.8,31.0,30.3,31.7);
  node["railway"="station"](29.8,31.0,30.3,31.7);
  node["railway"="subway_entrance"](29.8,31.0,30.3,31.7);
  node["highway"="bus_stop"](29.8,31.0,30.3,31.7);
  node["amenity"="bus_station"](29.8,31.0,30.3,31.7);
);
out body;
"""

osm_out = RAW_DIR / f"osm_overpass.json"
if not (osm_out.exists() and osm_out.stat().st_size > 10_000):
    r = Fetcher.post("https://overpass-api.de/api/interpreter", data={"data": OVERPASS_QL},
                     stealthy_headers=True, timeout=300)
    assert r.status == 200, f"Overpass failed: {r.status}"
    osm_out.write_bytes(r.body)

data = json.loads(osm_out.read_text(encoding='utf-8'))
feats = []
for el in data['elements']:
    if el.get('type') != 'node': continue
    tags = el.get('tags') or {}
    if not tags.get('name'): continue
    feats.append({'osm_id': el['id'], 'name': tags.get('name'),
                  'name_ar': tags.get('name:ar'), 'name_en': tags.get('name:en'),
                  'public_transport': tags.get('public_transport'),
                  'railway': tags.get('railway'), 'highway': tags.get('highway'),
                  'amenity': tags.get('amenity'), 'operator': tags.get('operator'),
                  'lat': el['lat'], 'lon': el['lon']})
osm_df = pd.DataFrame(feats)
osm_gdf = gpd.GeoDataFrame(osm_df, geometry=gpd.points_from_xy(osm_df.lon, osm_df.lat),
                           crs=CRS_MAP).to_crs(CRS_METERS)
osm_gdf.to_file(CLEAN_DIR / 'osm_features.geojson', driver='GeoJSON')
print(f"S8 OSM: {len(osm_gdf)} named transport nodes (cross-verification layer)")
```

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">LRT stations lack coordinates in Wikipedia. We match their names fuzzily against OSM feature names and copy the best-match coordinate (threshold 85). Unmatched stations keep null coords - they'll be handled explicitly in Stage 1.</p>

```python
# LRT coord sourcing — step A: per-station Overpass name search.
# LRT runs east from Adly Mansour toward the New Administrative Capital. Bbox is widened
# accordingly. For each Wikipedia station name we ask Overpass if OSM has a matching node.
LRT_BBOX = '29.5,31.0,31.0,32.5'

import time
import random
import json

def overpass_name(name, max_retries=5):
    ql = f'[out:json][timeout:600];(node["name"~"{name}",i]({LRT_BBOX});node["name:en"~"{name}",i]({LRT_BBOX}););out body 5;'

    for attempt in range(max_retries):
        r = Fetcher.post(
            'https://overpass-api.de/api/interpreter',
            data={'data': ql},
            stealthy_headers=True,
            timeout=600)

        if r.status == 200:
            return json.loads(r.body.decode('utf-8')).get('elements', [])

        if r.status == 429:
            wait = (2 ** attempt) + random.uniform(0.5, 1.5)
            print(f"Rate limited (429). Waiting {wait:.1f}s before retry...")
            time.sleep(wait)
        else:
            print(f"Overpass error {r.status}.")
            time.sleep(1)
            

    return []

lrt = pd.read_csv(CLEAN_DIR / 'lrt_stations.csv')
lrt_cache = RAW_DIR / f'osm_lrt_per_station.json'
if lrt_cache.exists() and lrt_cache.stat().st_size > 100:
    results = json.loads(lrt_cache.read_text())
else:
    results = {nm: overpass_name(nm) for nm in lrt.station_en.dropna().unique()}
    lrt_cache.write_text(json.dumps(results))
    time.sleep(0.5)

def _best_osm(hits):
    if not hits: return None, None, None
    ranked = sorted(hits, key=lambda e: (
        0 if e.get('tags',{}).get('railway') in ('station','halt','tram_stop','light_rail') else
        1 if e.get('tags',{}).get('public_transport') in ('station','platform') else 2))
    b = ranked[0]
    return b.get('lat'), b.get('lon'), b.get('tags',{}).get('name')

lats, lons, osm_names = [], [], []
for nm in lrt.station_en.fillna(''):
    la, lo, on = _best_osm(results.get(nm, []))
    lats.append(la); lons.append(lo); osm_names.append(on)
lrt['lat'] = lats; lrt['lon'] = lons; lrt['osm_hit_name'] = osm_names
osm_found = lrt[['lat','lon']].notna().all(axis=1).sum()
lrt.to_csv(CLEAN_DIR / 'lrt_stations.csv', index=False)
print(f"LRT step A (Overpass): {osm_found} / {len(lrt)} stations located")
```

<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Share Tech Mono',monospace;color:#58A6FF;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">⟐ CODE EXPLANATION</span>
 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
  For each unique district (keyed on <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">district + native_ar + status + governorate</code>) we collect whichever
  census years have a known population, fit a <strong>per-district OLS linear regression</strong> on
  <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">year → population</code>, and use it to fill every missing census year (1996, 2006, 2017, 2023).
  Districts with exactly one known point fall back to flat replication; districts with zero known points
  remain <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">NaN</code>. All predictions are clamped to ≥ 0. The imputed file overwrites
  <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">CleanedData/districts.csv</code>.
 </p>
</div>


```python
dist_path = CLEAN_DIR / "districts.csv"
dist_df   = pd.read_csv(dist_path)

GROUP_KEYS = ["district", "native_ar", "status", "governorate"]

filled_total = 0

for keys, grp in dist_df.groupby(GROUP_KEYS, sort=False):
    # Known (non-null, non-zero) observations for this district
    known = grp[grp["population"].notna() & (grp["population"] > 0)]

    if known.empty:
        continue   # nothing to extrapolate from — leave NaN

    known_years = known["year"].astype(int).tolist()
    known_pops  = known["population"].astype(float).tolist()

    # Rows that need filling
    missing_idx = grp.index[grp["population"].isna()]

    if missing_idx.empty:
        continue

    if len(known_years) >= 2:
        # Fit OLS on available years
        X   = np.array(known_years).reshape(-1, 1)
        y   = np.array(known_pops)
        reg = LinearRegression().fit(X, y)

        for idx in missing_idx:
            target_year = int(dist_df.at[idx, "year"])
            pred = reg.predict([[target_year]])[0]
            dist_df.at[idx, "population"] = max(0.0, round(pred, 1))
            filled_total += 1
    else:
        # Only 1 known point → flat replication
        flat_val = known_pops[0]
        for idx in missing_idx:
            dist_df.at[idx, "population"] = flat_val
            filled_total += 1

# Overwrite the CSV
dist_df.to_csv(dist_path, index=False)
print(f"✔ Imputed {filled_total:,} missing population cells · saved → {dist_path.name}")
print(dist_df["population"].isna().sum(), "cells still NaN (no known data for those districts)")
```

```python
lrt_path = CLEAN_DIR / "lrt_stations.csv"

lrt_df = pd.read_csv(lrt_path, dtype=str)

expected_cols = {"lat", "lon", "osm_hit_name"}
missing = expected_cols - set(lrt_df.columns)
if missing:
    raise KeyError(f"Missing columns in lrt_stations.csv: {missing}")

lrt_df["lat"] = pd.to_numeric(lrt_df["lat"], errors="coerce")
lrt_df["lon"] = pd.to_numeric(lrt_df["lon"], errors="coerce")
lrt_df["osm_hit_name"] = lrt_df["osm_hit_name"].fillna("Not available")

# Keep unknown coordinates as NaN. Downstream spatial analysis drops NaNs;
# replacing them with 0,0 creates false stations off the Cairo map.
missing_coords = lrt_df["lat"].isna() | lrt_df["lon"].isna()

lrt_df.to_csv(lrt_path, index=False)

print(f"Updated LRT stations: {missing_coords.sum():,} rows still have unknown coordinates and will be skipped spatially.")
```

```python

terms = pd.read_csv(PHASE1_CLEAN / 'cleaned_terminals.csv')
terms['geometry'] = terms['geometry'].apply(wkt.loads)
terms_gdf = gpd.GeoDataFrame(terms, geometry='geometry', crs=CRS_METERS)
terms_gdf['terminal_x'] = terms_gdf.geometry.centroid.x
terms_gdf['terminal_y'] = terms_gdf.geometry.centroid.y
print(f"Phase 1 terminals: {len(terms_gdf)}")

def to_gdf(df, x='lon', y='lat', src=''):
    d = df.copy()
    if x not in d.columns or y not in d.columns:
        # All-null geometry for downstream skipping
        g = gpd.GeoDataFrame(d, geometry=[None]*len(d), crs=CRS_METERS)
    else:
        d2 = d.dropna(subset=[x, y]).copy()
        if len(d2) == 0:
            g = gpd.GeoDataFrame(d, geometry=[None]*len(d), crs=CRS_METERS)
        else:
            g = gpd.GeoDataFrame(d2, geometry=gpd.points_from_xy(d2[x], d2[y]),
                                  crs=CRS_MAP).to_crs(CRS_METERS)
    g['source'] = src
    return g

metro = pd.read_csv(CLEAN_DIR / 'metro_stations.csv')
lrt   = pd.read_csv(CLEAN_DIR / 'lrt_stations.csv')
brt   = pd.read_csv(CLEAN_DIR / 'brt_stations.csv')
osm   = gpd.read_file(CLEAN_DIR / 'osm_features.geojson').to_crs(CRS_METERS)
osm['source'] = 'S8_osm'

metro_g = to_gdf(metro, src='S3_metro')
lrt_g   = to_gdf(lrt,   src='S4_lrt')
brt_g   = to_gdf(brt,   src='S5_brt')

print(f"  metro stations w/coords: {metro_g.geometry.notna().sum()} / {len(metro_g)}")
print(f"  lrt stations   w/coords: {lrt_g.geometry.notna().sum()} / {len(lrt_g)}")
print(f"  brt stations   w/coords: {brt_g.geometry.notna().sum()} / {len(brt_g)}")
print(f"  osm features:            {len(osm)}")
```

 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">For every scraped station with a non-null geometry, find the nearest Phase 1 terminal. Pairs under ~200 m are likely the same location; over 1 km means name-based matching must take over in Stages 3/4.</p>

```python
# Stage 1 — spatial KNN: nearest Phase 1 terminal for every scraped station with coords.
def knn_nearest(scraped_g, terminals_gdf, k=1):
    g = scraped_g[scraped_g.geometry.notna()].copy()
    if len(g) == 0:
        return pd.DataFrame(columns=['source','scraped_name','nn_terminal_name','nn_terminal_id','dist_m'])
    T = terminals_gdf[['terminal_x','terminal_y']].to_numpy()
    S = np.column_stack([g.geometry.x, g.geometry.y])
    nn = NearestNeighbors(n_neighbors=k).fit(T)
    dist, idx = nn.kneighbors(S)
    name_col = 'station_en' if 'station_en' in g.columns else 'name'
    return pd.DataFrame({
        'source':           g['source'].values,
        'scraped_name':     g[name_col].values,
        'nn_terminal_name': terminals_gdf.iloc[idx[:,0]]['Terminal_Name'].values,
        'nn_terminal_id':   terminals_gdf.iloc[idx[:,0]]['Terminal_ID'].values,
        'dist_m':           dist[:,0]})

stage1_metro = knn_nearest(metro_g, terms_gdf)
stage1_lrt   = knn_nearest(lrt_g,   terms_gdf)
stage1_brt   = knn_nearest(brt_g,   terms_gdf)
stage1_osm   = knn_nearest(osm,     terms_gdf)
stage1 = pd.concat([stage1_metro, stage1_lrt, stage1_brt, stage1_osm], ignore_index=True)
```

```python
# Stage 2 — OSM cross-verification: flag Wikipedia stations whose nearest OSM node is >50m away.
def crossverify_osm(scraped_g, osm_gdf):
    g = scraped_g[scraped_g.geometry.notna()].copy()
    if len(g) == 0 or len(osm_gdf) == 0:
        return pd.DataFrame()
    O = np.column_stack([osm_gdf.geometry.x, osm_gdf.geometry.y])
    S = np.column_stack([g.geometry.x, g.geometry.y])
    nn = NearestNeighbors(n_neighbors=1).fit(O)
    dist, idx = nn.kneighbors(S)
    name_col = 'station_en' if 'station_en' in g.columns else 'name'
    return pd.DataFrame({'source': g['source'].values,
                         'scraped_name': g[name_col].values,
                         'nearest_osm_name': osm_gdf.iloc[idx[:,0]]['name'].values,
                         'osm_distance_m': dist[:,0],
                         'flagged_discrepancy': dist[:,0] > 50})

stage2 = pd.concat([crossverify_osm(metro_g, osm),
                    crossverify_osm(lrt_g, osm),
                    crossverify_osm(brt_g, osm)], ignore_index=True)
print(f"Stage 2 · OSM verify: {len(stage2)} pairs · {(stage2.osm_distance_m<=50).sum()} within 50m · {stage2.flagged_discrepancy.sum()} flagged")
```

```python
# Stage 3 — RapidFuzz token_sort: English-name transliteration variants.
# Good: 'Al-Shohadaa' ↔ 'Al Shohada' · Bad: cross-script pairs (handled by Stage 4).
def fuzzy_match(scraped_names, terminal_names, threshold=88):
    out = []
    for nm in scraped_names:
        if not isinstance(nm, str) or not nm.strip(): continue
        m = process.extractOne(nm, terminal_names, scorer=fuzz.token_sort_ratio)
        out.append({'scraped_name': nm,
                    'matched_terminal': m[0] if m and m[1] >= threshold else None,
                    'fuzzy_score': int(m[1]) if m else 0})
    return pd.DataFrame(out)

all_scraped = [s for s in (list(metro['station_en']) + list(lrt['station_en']) + list(brt['station_en']))
               if isinstance(s, str) and s.strip()]
terminal_names = terms_gdf['Terminal_Name'].dropna().unique().tolist()
stage3 = fuzzy_match(all_scraped, terminal_names, threshold=88)
print(f"Stage 3 · fuzzy: {len(stage3)} pairs · {stage3.matched_terminal.notna().sum()} matched at τ=88")
```

 <span style="font-family:'Share Tech Mono',monospace;color:#7B2D8E;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">METHODOLOGY ENHANCEMENT</span>
 <p style="margin:10px 0 0;color:#C9D1D9;line-height:1.65;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;"><strong>AI Technique 1 · Multilingual similarity scoring.</strong> RapidFuzz operates on character n-grams and cannot bridge writing systems - <em>Heliopolis</em> and <em>مصر الجديدة</em> share zero characters. We embed scraped station names and scraped Phase 1 terminal names with <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">paraphrase-multilingual-MiniLM-L12-v2</code>, then use cosine similarity only as a score. The accepted rule also requires spatial support: <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">semantic_score >= 0.75</code> and nearest Phase 1 terminal within <code style="white-space:normal;overflow-wrap:anywhere;word-break:break-word;max-width:100%;">2 km</code>. All names and coordinates come from the scraped sources; the embedding model only scores similarity between them.</p>
</div>


<div style="background:rgba(88,166,255,0.06);border-left:3px solid #58A6FF;padding:14px 20px;border-radius:0 6px 6px 0;margin:16px 0;width:calc(100% - 32px);max-width:calc(100% - 32px);box-sizing:border-box;overflow-x:auto;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">
 <span style="font-family:'Share Tech Mono',monospace;color:#58A6FF;font-size:.85em;letter-spacing:.15em;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">CODE EXPLANATION</span>
 <p style="margin:8px 0 0;color:#C9D1D9;line-height:1.6;max-width:100%;box-sizing:border-box;overflow-wrap:anywhere;word-break:break-word;white-space:normal;">Build one candidate row per scraped Metro/LRT/BRT station. The model embeds the scraped names and Phase 1 terminal names, selects the closest match, and records the cosine similarity score. A candidate is accepted only if the score and spatial guard both pass; review-needed rows stay in the audit table.</p>
</div>


```python
# Stage 4 — AI Technique 1: multilingual similarity scoring, not data generation.
SEMANTIC_THRESHOLD = 0.75
SPATIAL_THRESHOLD_M = 2000

model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

def _combined(en, ar):
    en = '' if pd.isna(en) else str(en)
    ar = '' if pd.isna(ar) else str(ar)
    return (en + ' ' + ar).strip()

terms_gdf['combined'] = terms_gdf.apply(lambda r: _combined(r['Terminal_Name'], r['Terminal_Name_AR']), axis=1)
term_emb = model.encode(terms_gdf['combined'].tolist(), batch_size=32, show_progress_bar=False)

def _unify(df, src):
    d = df.copy()
    if 'station_ar' not in d.columns:
        d['station_ar'] = ''
    return d[['station_en','station_ar']].assign(source=src)

scraped_df = pd.concat([
    _unify(metro, 'S3_metro'),
    _unify(lrt,   'S4_lrt'),
    _unify(brt,   'S5_brt'),
], ignore_index=True)
scraped_df['combined'] = scraped_df.apply(lambda r: _combined(r.station_en, r.station_ar), axis=1)
scraped_emb = model.encode(scraped_df['combined'].tolist(), batch_size=32, show_progress_bar=False)

sim = cosine_similarity(scraped_emb, term_emb)
best_idx = sim.argmax(axis=1)
best_score = sim.max(axis=1)

stage4 = scraped_df.copy()
stage4['matched_terminal_id'] = terms_gdf.iloc[best_idx]['Terminal_ID'].values
stage4['matched_terminal_name'] = terms_gdf.iloc[best_idx]['Terminal_Name'].values
stage4['semantic_score'] = best_score

# Stage 1 KNN supplies the spatial guard; AI similarity alone is never enough.
_stage1_all = pd.concat([stage1_metro, stage1_lrt, stage1_brt], ignore_index=True)
_dist_lookup = _stage1_all.groupby('scraped_name')['dist_m'].min().to_dict()
stage4['nearest_terminal_m'] = stage4['station_en'].map(_dist_lookup).fillna(float('inf'))
stage4['fuzzy_score'] = stage4['station_en'].map(
    stage3.set_index('scraped_name')['fuzzy_score'].to_dict()).fillna(0).astype(int)
stage4['semantic_ok'] = (
    (stage4['semantic_score'] >= SEMANTIC_THRESHOLD) &
    (stage4['nearest_terminal_m'] < SPATIAL_THRESHOLD_M)
)
stage4['match_status'] = np.where(stage4['semantic_ok'], 'accepted', 'review_needed')
stage4['match_rule'] = np.where(
    stage4['semantic_ok'],
    f'semantic_score>={SEMANTIC_THRESHOLD} and nearest_terminal<{SPATIAL_THRESHOLD_M}m',
    'below semantic or spatial threshold'
)

upgraded = ((stage4.semantic_ok) & (stage4.fuzzy_score < 88)).sum()
print(
    f"Stage 4 · AI similarity scoring: {len(stage4)} candidates · "
    f"{stage4.semantic_ok.sum()} accepted · {upgraded} accepted beyond fuzzy · no generated data"
)
```

Counts are overlapping diagnostics, not a strict funnel, because one station can have spatial, fuzzy, and AI-similarity evidence at the same time.</p>

```python
# Integration diagnostics — overlapping evidence counts by source.
def _yield(scraped_df, s1, s3_sub, s4_sub):
    n_in = len(scraped_df)
    accepted = int(s4_sub.semantic_ok.sum()) if len(s4_sub) else 0
    return (n_in,
            int((s1.dist_m < 200).sum()) if len(s1) else 0,
            int(s3_sub.matched_terminal.notna().sum()) if len(s3_sub) else 0,
            accepted,
            max(0, n_in - accepted))

yields = {
    'metro': _yield(metro, stage1_metro, stage3[stage3.scraped_name.isin(metro.station_en)], stage4[stage4.source=='S3_metro']),
    'lrt':   _yield(lrt,   stage1_lrt,   stage3[stage3.scraped_name.isin(lrt.station_en)],   stage4[stage4.source=='S4_lrt']),
    'brt':   _yield(brt,   stage1_brt,   stage3[stage3.scraped_name.isin(brt.station_en)],   stage4[stage4.source=='S5_brt']),
    'osm':   _yield(osm,   stage1_osm,   pd.DataFrame(), pd.DataFrame()),
}

labels = ['S3 metro','S4 LRT','S5 BRT','S8 OSM',
          'Spatial evidence: KNN<200m','Name evidence: fuzzy>=88',
          f'Accepted: score>={SEMANTIC_THRESHOLD} + distance<{SPATIAL_THRESHOLD_M}m',
          'Review needed / not terminal-matched']
src_idx, tgt_idx, vals = [], [], []
for i, (_, v) in enumerate(yields.items()):
    n_in, s1, s3, accepted, review = v
    if s1: src_idx.append(i); tgt_idx.append(4); vals.append(s1)
    if s3: src_idx.append(i); tgt_idx.append(5); vals.append(s3)
    if accepted: src_idx.append(i); tgt_idx.append(6); vals.append(accepted)
    if review: src_idx.append(i); tgt_idx.append(7); vals.append(review)

fig = go.Figure(go.Sankey(
    node=dict(label=labels, pad=18, thickness=20,
              color=['#58A6FF','#2A9D8F','#E86F51','#4ECDC4','#E9C46A','#F4A261','#7B2D8E','#888']),
    link=dict(source=src_idx, target=tgt_idx, value=vals, color='rgba(88,166,255,0.25)')))
fig.update_layout(template='plotly_dark', paper_bgcolor='#0D1117',
                  title=dict(text='Integration diagnostics · scraped data with AI similarity scoring',
                             font=dict(color='#E9C46A', size=16)),
                  font=dict(color='#C9D1D9'), height=560, width=1150)
fig.show()
```

```python
# Persist final integration outputs: matched_pairs.csv + Phase2_integrated.geoparquet.
matched_pairs = stage4[[
    'source','station_en','station_ar','matched_terminal_id','matched_terminal_name',
    'fuzzy_score','semantic_score','nearest_terminal_m','semantic_ok','match_status','match_rule'
]].copy()
matched_pairs.to_csv(INTEGRATED_DIR / 'matched_pairs.csv', index=False)

def _to_geoparquet_row(df, src, line=None, year=None):
    d = df.copy()
    if {'lat','lon'}.issubset(d.columns) and d[['lat','lon']].notna().any(axis=None):
        dd = d.dropna(subset=['lat','lon']).copy()
        g = gpd.GeoDataFrame(dd, geometry=gpd.points_from_xy(dd.lon, dd.lat), crs=CRS_MAP).to_crs(CRS_METERS)
    else:
        g = gpd.GeoDataFrame(d, geometry=[None]*len(d), crs=CRS_METERS)
    g['source_label'] = src
    if line:
        g['line'] = line
    if year:
        g['opening_year'] = year
    return g

combined = gpd.GeoDataFrame(pd.concat([
    _to_geoparquet_row(metro, 'S3_metro')[['source_label','station_en','station_ar','line','phase','is_interchange','geometry']],
    _to_geoparquet_row(lrt,   'S4_lrt', line='LRT', year=2022)[['source_label','station_en','line','opening_year','geometry']],
    _to_geoparquet_row(brt,   'S5_brt', line='BRT', year=2025)[['source_label','station_en','station_ar','line','opening_year','geometry']],
    osm[['name','railway','public_transport','amenity','highway','geometry']].rename(columns={'name':'station_en'}).assign(source_label='S8_osm'),
], ignore_index=True), geometry='geometry', crs=CRS_METERS)
combined.to_parquet(INTEGRATED_DIR / 'Phase2_integrated.geoparquet')
print(
    f"Integration outputs · matched_pairs.csv ({len(matched_pairs)} candidates, "
    f"{matched_pairs.semantic_ok.sum()} accepted) · "
    f"Phase2_integrated.geoparquet ({len(combined)} map features)"
)
```
